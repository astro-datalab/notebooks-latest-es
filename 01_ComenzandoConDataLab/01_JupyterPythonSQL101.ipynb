# SQL - Lenguaje de consulta estructurado

<a class="anchor" id="querystructure"></a>
## SQL es una forma de describir a una base de datos lo que quieres de ella.
Plantilla general para una consulta sencilla escrita en SQL
```
SELECT algo
FROM base de datos.tabla
WHERE restricciones
```

El "algo" suele ser una lista de nombres de columnas, por ejemplo `SELECT ra,dec`

Una "base de datos" tiene un nombre y normalmente contiene varias "tablas".

P.ej. para acceder a las columnas de la base de datos `gaia_dr3`, tabla `gaia_source`, se podr铆a solicitar
`gaia_dr3.gaia_source`

Las "restricciones" son filtros o calificadores de la consulta, por ejemplo:

`WHERE ra>90`

O combine varias restricciones:

```
WHERE ra>90 AND ra<100
AND dism>40 AND dism<42
```

<a class="anchor" id="limit"></a>
## LIMITATE
Si agrega un `LIMIT 100` (o cualquier n煤mero peque帽o) a su consulta, la base de datos devolver谩 solo las primeras 100 filas de resultados. Esto suele ser r谩pido y puedes modificar tu consulta muchas veces de esta manera hasta que lo hagas bien.

**Excelente consejo al probar consultas: LIMITATE.**

La plantilla se convierte en:
```
SELECT algo
FROM base de datos.tabla
WHERE restricciones
LIMIT 100
```

## Nuestra primera consulta de Data Lab
Preguntemos por las velocidades radiales de todas las estrellas en Gaia DR3 alrededor de la posici贸n del LMC

````python
# coordenadas de la LMC
# ra = 81.
# diciembre = -69,8

query = """
SELECT radial_velocity
FROM gaia_dr3.gaia_source
WHERE ra>78 and ra<84
  AND dec>-73 and dec<-67
  AND radial_velocity != 'nan'
  AND abs(radial_velocity)>0.
  """

print(query)
````

Ahora env铆emos la consulta a la base de datos del Data Lab.

````python
result = qc.query(sql=query)
````

# Si eres demasiado suave para estar sentado: consultas as铆ncronas
Una consulta que podr铆a ejecutarse potencialmente por m谩s tiempo que su capacidad de atenci贸n... En ese caso, simplemente env铆ela en modo as铆ncrono y recupere los resultados m谩s tarde, cuando finalice la consulta (y cuando haya tomado su merecido caf茅):

````python
query = """SELECT ra,dec FROM smash_dr1.object 
WHERE fieldid = 169 AND depthflag > 1 
AND ndetr > 3 AND ndetg > 3 
AND abs(sharp) < 0.5 
AND gmag BETWEEN 9 AND 25 
AND (gmag-rmag) BETWEEN -0.4 AND 0.4"""
````

Las consultas asincr贸nicas obtienen un id de trabajo (jobid) al enviarse, por ejemplo:

```
jobid = qc.query(adql=query,async_=True)
print (oferta de trabajo)
q5qy6ujsu9cnygcp
```

Luego puede verificar peri贸dicamente el estado (status) de la consulta enviando la cadena `jobid` al m茅todo `status`:

```
qc.status(jobid)
'EXECUTING'
```

Si repite la solicitud de "estado" un poco m谩s tarde, eventualmente cambiar谩 a "COMPLETADO", momento en el cual podr谩 recuperar el resultado:

```
resultado = qc.results(jobid)
```

**Afortunadamente, existe un mecanismo incorporado para realizar la verificaci贸n peri贸dica por usted.** Tambi茅n recupera los resultados de su consulta al final. La firma es:

```
resultado = qc.query(sql=query,async_=True,wait=True,poll=5,verbose=1)
```

donde `poll` establece el per铆odo de survey (en segundos).

````python
resultado = qc.query(adql=query,async_=True,wait=True,poll=5,verbose=1)
````

Convierta al marco de datos de Pandas, como antes:

````python
df = convert(resultado)
````

Utilice el marco de datos como desee, por ejemplo, as铆 para graficar RA y Dec:

````python
plt.figure(figsize=(5,5))
plt.hexbin(df['ra'],df['dec'])
plt.xlabel('AR')
plt.ylabel('Dec')
````
