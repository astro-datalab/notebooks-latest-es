# Información básica sobre las tablas de la base de datos

### ¿Qué conjuntos de datos están disponibles?
El queryClient tiene un método `schema` que proporciona información sobre las bases de datos, tablas y columnas disponibles. Si llamamos a `qc.schema()` con un primer argumento vacío, obtendremos información sobre los conjuntos de datos disponibles y una descripción de una línea para la mayoría de ellos.

```python
print(qc.schema())
```

### Obtener la lista de tablas
Si llamamos a `qc.schema()` con el nombre de un conjunto de datos específico, veremos qué tablas están disponibles para ese conjunto de datos. Esto es lo que está disponible para SMASH DR1:

```python
print(qc.schema('smash_dr1'))
```

### Obtener la lista de columnas
También podemos usar `qc.schema()` para obtener los nombres y descripciones de las columnas de una tabla específica. Esto es lo que está disponible para la tabla de objetos de SMASH DR1. (Nótese que no todos los conjuntos de datos tienen descripciones para cada columna).

```python
print(qc.schema('smash_dr1.object'))
```

### Obtener estadísticas de tablas
A menudo, querrá obtener información básica sobre una tabla determinada, por ejemplo, el número de filas. Podría verse tentado a usar `SELECT COUNT(*)` en una consulta para esto, pero esto puede ser lento con conjuntos de datos grandes y usualmente, no es necesario. En su lugar, la base de datos especial `tbl_stat` contiene esta información para cada conjunto de datos. Consultemos esta tabla:

```python
#query = "SELECT COUNT(ra) FROM phat_v2.phot_mod" # SLOW
query = "SELECT * FROM tbl_stat WHERE schema='smash_dr1' AND tbl_name='object'" # Obtiene estadísticas útiles, rápidamente
```

```python
info = qc.query(sql=query) # Por omisión, el resultado es una cadena con formato CSV.
```

# Una consulta simple para datos de catálogo
En la sección anterior, ya vimos una consulta básica a la base de datos `tbl_stats`. Aquí recuperaremos las primeras 10 filas de la tabla `smash_dr1.object`:

```python
query = "SELECT * FROM smash_dr1.object LIMIT 10"
result = qc.query(sql=query) # Por omisión, el resultado es una cadena con formato CSV.
```

Nótese que por omisión, el resultado se devuelve como una cadena larga con el aspecto de un archivo CSV: 

```python
print(type(result))
print(result[:1000]) # Imprime solo los primeros 1000 caracteres de la cadena que retorna
```

Generalmente, queremos convertir esta cadena en una tabla o matriz, es decir, en un formato que permita el procesamiento numérico. El módulo `helpers` de Data Lab lo facilita. Aquí convertimos la cadena resultante en un marco de datos de Pandas:

```python
df = convert(result,'pandas')
df.head()
```

Considere que, para mayor comodidad, se puede solicitar que el resultado esté en otros formatos, incluyendo un marco de datos de Pandas:

```python
query = "SELECT * FROM smash_dr1.object LIMIT 10"
df = qc.query(sql=query,fmt='pandas')
df.head()
```

# Buscar y descargar un recorte de imagen
Usar el servicio de recorte de imagen es un proceso de dos pasos. Primero, especificamos una posición (RA y Dec) y el tamaño de la imagen que queremos recuperar. Segundo, buscamos en el servicio SIA todas las imágenes que se superponen a ese punto en el cielo:

```python
# NGC 288 (un cúmulo globular)
ra = 13.19     # en grados decimales
dec = -26.59   # en grados decimales

fov = 13/60  # Campo de visión del recorte de imagen (en grados; aquí 13 arminutos = 0,22 grados)

imgTable = svc.search((ra,dec), (fov/np.cos(dec*np.pi/180), fov), verbosity=2).to_table()  # usa corrección de declinación 
```

El resultado es una tabla VOTable que convertimos en una tabla astropy sobre la marcha. La tabla contiene numerosas columnas de metadatos que describen los parámetros de cada imagen, incluyendo la URL del recorte. Sin embargo, cabe destacar que las imágenes sobre la calidad de datos, como máscaras (`masks`) o mapas de peso (`weight`), también pueden aparecer en la lista:

```python
print(type(imgTable))
imgTable
```

El siguiente paso es identificar la imagen deseada de la lista de imágenes disponibles. Aquí, limitaremos la lista a pilas de imágenes de banda g y seleccionaremos el tipo de producto `image` (en lugar de `weight` o `mask`).

```python
# selección booleana para imágenes en la banda g solamente
sel0 = imgTable['obs_bandpass'] == 'g'

# añadir lógicamente selecciones para los tipos de producto 'Stack' e 'image'
sel = sel0 & ((imgTable['proctype'] == 'Stack') & (imgTable['prodtype'] == 'image')) # selección básica

# filtrar la tabla de resultados anterior
Table = imgTable[sel] # seleccionar
Table
```

el último paso es extraer la URL y descargar la imagen:

```python
row = Table[0]
url = row['access_url'] # obtener la URL de descarga
print(url)
```

```python
gimage = io.fits.getdata(utils.data.download_file(url,cache=True,show_progress=False,timeout=120))
```

Crearemos una visualización rápida aquí:

```python
fig = plt.subplots(figsize=(10,10))
plt.imshow(np.arcsinh(gimage),cmap='gray',norm=matplotlib.colors.Normalize(vmin=0))
plt.axis('off');
```

## Creémos una imagen a color.
Escribiremos una función rápida para realizar la consulta SIA y la selección de imágenes:

```python
# Una función para descargar las imágenes apiladas más profundas disponibles
def download_deep_stack(ra,dec,fov=0.1,band='g'):
    imgTable = svc.search((ra,dec), (fov/np.cos(dec*np.pi/180), fov), verbosity=2).to_table()
    print("La lista completa de imágenes contiene", len(imgTable), "entradas")

    # aplicar los criterios para seleccionar las imágenes
    sel0 = imgTable['obs_bandpass'] == band
    sel = sel0 & ((imgTable['proctype'] == 'Stack') & (imgTable['prodtype'] == 'image')) # selección básica
    Table = imgTable[sel] # select

    # Si más de una imagen cumple con nuestros criterios
    if (len(Table)>0):
        row = Table[np.argmax(Table['exptime'].data.data.astype('float'))] # seleccionar la imagen con el mayor tiempo de exposición
        url = row['access_url'] # obtener la URL de descarga
        print ('Descargando la imagen apilada más profunda...')
        image = io.fits.getdata(utils.data.download_file(url,cache=True,show_progress=False,timeout=120))
        print(url)

    # Si ninguna imagen cumple con nuestros criterios
    else:
        print ('No hay ninguna imagen disponible.')
        image = None
        
    print()
    return image
```

Con esta función, podemos descargar rápidamente dos bandas más (bandas `r` e `i`) y combinarlas con la imagen de banda g para crear una imagen en falso color:

```python
rimage = download_deep_stack(ra,dec,fov,band='r')
iimage = download_deep_stack(ra,dec,fov,band='i')
```

Ahora podemos combinar imágenes en tres bandas:

```python
color_image = make_lupton_rgb(iimage, rimage, gimage, stretch=100, Q=0.1)
```

Y graficar la imagen resultante en color falso:

```python
fig = plt.figure(figsize=(10,10))
plt.imshow(color_image)
plt.axis('off');
```
