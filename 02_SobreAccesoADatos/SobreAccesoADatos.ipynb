# Distribución de magnitud
En este ejemplo, consultamos la base de datos AllWISE (tabla `source`) para obtener las magnitudes medidas en tres bandas de 10 mil fuentes con alta relación señal/ruido (SNR).

```python
query =\
"""
select w1mpro,w2mpro,w3mpro
from allwise.source
where w1snr>30 and w2snr>30 and w3snr>30
limit 10000
"""
res = qc.query(sql=query,format='csv')
```

Convirtamos el resultado de la consulta a un formato más conveniente (en este caso, un dataframe de Pandas).

```python
df = convert(res,'pandas')
```

Gráficaremos las distribuciones de las magnitudes W1, W2 y W3 de WISE como histogramas.

```python
colors = list('rgb')
fig, ax = plt.subplots(1,1,figsize=(7,5))
for j in (1,2,3):
    ax.hist(df['w{:d}mpro'.format(j)],bins=40,range=(6,10),histtype='step',lw=3,color=colors[j-1],alpha=0.4,label='W{:d}'.format(j+1))
ax.set_xlabel('mag (Vega)')
ax.set_ylabel('Histograma')
ax.legend(loc='upper left',frameon=False)
```

# Diagrama color-magnitud
En este ejemplo, consultamos la fotometría de estrellas alrededor de la ubicación conocida de la galaxia enana Hydra II y generamos un diagrama color-magnitud (DCM) de ellas.

Definamos la cadena de consulta. Solicitamos objetos puntuales `(abs(sharp<0.5))` en el campo número `169`, detectados en imágenes de larga exposición (`depthflag>1`), con magnitudes `g` entre `9` y `25`, alrededor de una posición central `ra0/dec0`, dentro de un radio de `0.1` grados.

```python
field, ra0, dec0, radius = 169, 185.410559533, -31.9760034474, 0.1 # ra0,dec0,radio todos en grados decimales
query = """
SELECT ra,dec,gmag,rmag,imag FROM smash_dr1.object
  WHERE fieldid = {:d}
    AND depthflag > 1
    AND abs(sharp) < 0.5
    AND gmag BETWEEN 9 AND 25
    AND q3c_radial_query(ra,dec,{:f},{:f},{:f})
""".format(field,ra0,dec0,radius)
```

Envíe la consulta y conviértala a Pandas:

```python
respuesta = qc.query(sql=query,format='csv')
df = convert(respuesta,'pandas')
print(df.head())
```

Calcule una columna de color (en el mismo marco de datos):

```python
df['g_r'] = df['gmag'] - df['rmag']
print(df.head())
```

Grafique un diagrama color-magnitud:

```python
fig, ax = plt.subplots(figsize=(5,5))
x = df['g_r']
y = df['gmag']
ax.scatter(x,y,marker='.',s=10, alpha=0.8)
ax.set_xlabel(x.name)
ax.set_ylabel(y.name)
ax.set_xlim((-1,2))
ax.set_ylim((25.2,14))
title = ax.set_title('{:d} objetos'.format(len(df)))
```

# Curva de luz
Algunos catálogos tienen datos multi-época, como el survey SMASH. Consultaremos las magnitudes y sus errores de todas las mediciones de una estrella, cuyo ID conocemos.

```python
# ID único de una estrella RR Lyrae en la galaxia enana Hidra II
objID = '169.429960'

# Seleccionar columnas: RA, Dec, fecha juliana modificada, magnitud calibrada, incertidumbres, banda de filtro
# Nota: Esta tabla de la base de datos codifica los valores 'sin medición' como 99,99
# Ordenar las filas devueltas en orden ascendente de fecha juliana modificada
query = """SELECT ra,dec,mjd,cmag,cerr,filter
        FROM smash_dr1.source
        WHERE id='{:s}' AND cmag<99
        ORDER BY mjd ASC""".format(objID)
result = qc.query(sql=query) # Por omisión, el resultado es una cadena en formato CSV
```

Convertir a Pandas para mayor comodidad

```python
df = convert(result,'pandas') # Convertir el resultado de la consulta a un dataframe de Pandas
print(df.head())
```

Grafiquemos la curva de luz en función de la fecha

```python
fig, ax = plt.subplots(figsize=(8,6))
co = (df['filter']=='g')
start = 16
jd0 = 57110
ax.errorbar(df['mjd'][co][start:]-jd0,df['cmag'][co][start:],yerr=df['cerr'][co][start:],marker='.',ms=8,ls='none',color='g',lw=1,alpha=0.5,label='')
ax.set_xlabel('Fecha Juliana Modificada - {:d}'.format(jd0))
ax.set_ylabel('Magnitud g')
```
