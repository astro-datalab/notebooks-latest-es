# Crear un diagrama color-color

Las muchas bandas fotométricas de S-PLUS pueden usarse para producir diagramas color-color que son útiles para medir propiedades como la metalicidad o la gravedad superficial de una estrella.

(g-F515) vs. (g-r) es sensible a la gravedad superficial y puede usarse para separar estrellas enanas de estrellas gigantes. El locus principal de puntos corresponde a las enanas, mientras que las gigantes rojas se encuentran en la parte superior derecha. Consulta [Majewski et al. (2000)](http://adsabs.harvard.edu/abs/2000AJ....120.2550M) para más detalles sobre este método.

```python
fig = plt.figure(figsize=(7,6))
bright = (R['g_auto']<20)
plt.scatter(R['g_auto'][bright]-R['r_auto'][bright], R['g_auto'][bright]-R['f515_auto'][bright],c=R['g_auto'][bright],s=10)
plt.xlim(-1,2)
plt.ylim(-0.5,0.8)
plt.xlabel('g-r')
plt.ylabel('g-F515')
plt.show()
```

El diagrama color-color (F378-F410) vs. (F515-F861) es similar y también puede usarse para derivar la gravedad superficial. Consulta [Cenarro et al. (2018)](http://adsabs.harvard.edu/abs/2018arXiv180402667C) para más detalles.

```python
fig = plt.figure(figsize=(7,6))
bright = (R['g_auto']<19)
plt.scatter(R['f515_auto'][bright]-R['f861_auto'][bright], R['f378_auto'][bright]-R['f410_auto'][bright],c=R['g_auto'][bright],s=10)
plt.xlim(-1,3)
plt.ylim(2.0,0.0)
plt.xlabel('F515-F861')
plt.ylabel('F378-F410')
cbar = plt.colorbar()
cbar.set_label('magnitud g')
```

# Crear Distribuciones de Energía Espectral (SED)

Una de las grandes ventajas del conjunto de datos de S-PLUS es el gran número de filtros de banda ancha y angosta que pueden usarse para estudiar las propiedades de los objetos. Aquí mostramos cómo graficar las Distribuciones de Energía Espectral (SEDs por sus siglas en inglés) de los objetos. Estas pueden compararse luego con modelos espectrales para restringir propiedades como Teff, logg y metalicidad (en el caso de estrellas), o el tipo de galaxia y su corrimiento al rojo (redshift) en el caso de galaxias. (Consulta [Cenarro et al. 2018](http://adsabs.harvard.edu/abs/2018arXiv180402667C) para algunos ejemplos).

```python
fig = plt.figure(figsize=(16,16))
font = {'family' : 'DejaVu Sans',
        'weight' : 'normal',
        'size'   : 12}
plt.rc('font',**font)
bands = ['ujava','f378','f395','f410','f430','g','f515','r','f660','i','f861','z']
wave = np.array([3536.,3733.,3941.,4095.,4293.,4780.,5134.,6267.,6614.,7684.,8608.,8956.])
for j in range(9):
    ax = fig.add_subplot(3,3,j+1)
    mag = []
    err = []
    for b in bands:
        mag.append(R[b+'_auto'][j])
        err.append(R['s2n_'+b+'_auto'][j])
    mag = np.array(mag)
    err = 1.087/np.array(err)
    gd = (mag < 50)
    ax.plot(wave[gd],mag[gd])
    #ax.scatter(wave[gd],mag[gd])
    #ax.errorbar(wave[gd],mag[gd],yerr=err[gd], fmt='o', color='black',
    #         ecolor='lightgray', elinewidth=3, capsize=0);
    ax.errorbar(wave[gd],mag[gd],yerr=err[gd], fmt='.k',markersize=10)
    ax.set_xlim(3400,9100)
    ax.set_ylim(np.max(mag[gd])+1.0,np.min(mag[gd])-1.0)
    ax.set_xlabel('Longitud de Onda (Å)')
    ax.set_ylabel('Magnitud')
    ax.text(4100.0,np.min(mag[gd])-1.0+0.07*(np.max(mag[gd])-np.min(mag[gd])+2.0),R['id'][j]) 
```

# Recortes de Imágenes

Vamos a obtener algunos recortes de imágenes y crear una imagen en tres colores.

Primero definimos algunas funciones auxiliares.

```python
# Configurando el SIA (Acceso Simple a Imágenes, por sus siglas en inglés)
DEF_ACCESS_URL = "https://datalab.noirlab.edu/sia/splus_dr1"
svc = sia.SIAService(DEF_ACCESS_URL)

# Una pequeña función para descargar las imágenes apiladas más profundas
def download_deepest_images(ra,dec,fov=0.1,bands=list('GRI')):
    imgTable = svc.search((ra,dec), (fov/np.cos(dec*np.pi/180), fov), verbosity=2).to_table()
    print("La lista completa de imágenes contiene {:d} entradas.".format(len(imgTable)))
    sel0 = (imgTable['proctype'] == 'Stack') & (imgTable['prodtype'] == 'image') # selección básica
    images = []
    for band in bands:
        print("Banda {:s}: ".format(band)) #, end=''
        sel = sel0 & (imgTable['obs_bandpass'] == band) # añadir 'banda' a la selección
        Table = imgTable[sel] # seleccionar
        row = Table[np.argmax(Table['exptime'].data.data.astype('float'))] # elegir la imagen con mayor tiempo de exposición
        url = row['access_url'] # obtener la URL de descarga
        print('descargando la imagen apilada más profunda...')
        img = io.fits.getdata(utils.data.download_file(url,cache=True,show_progress=False,timeout=120)) # .decode() porque en Python 3 url es de tipo "byte" y getdata() espera tipo "string"
        images.append(img)
        
    print("Se descargaron {:d} imágenes.".format(len(images)))
    return images

# Graficador de imágenes en múltiples paneles
def plot_images(images,geo=None,panelsize=7,titles=list('gri'),cmap=plt.cm.gray_r):
    if geo is None:
        geo = (2,2)
        
    fig = plt.figure(figsize=(geo[0]*panelsize,geo[1]*panelsize))
    for j,img in enumerate(images):
        ax = fig.add_subplot(geo[1],geo[0],j+1)
        ax.imshow(img,origin='lower',interpolation='none',cmap=cmap,norm=plt.mpl.colors.PowerNorm(0.1))
        ax.set_title('{:s}'.format(titles[j]))
        plt.axis('off')
```

Ahora descargamos las imágenes apiladas más profundas en tres bandas y las combinamos para crear una imagen en color con 3 bandas.

```python
bands = ['F515','R','I']
idx = 1
ra = 315.15
dec = -0.8
images = download_deepest_images(ra,dec, fov=0.1, bands=bands) # FOV in deg
```

```python
images2 = [im-np.median(im) for im in images] # restar la mediana de todas las imágenes para mejorar el escalado
images2 += [make_lupton_rgb(*images2[::-1],stretch=0.5)] # añadir una imagen compuesta en falso color con 3 bandas
plot_images(images2,titles=bands+['Imagen en falso color (3 bandas)'])
```
