# Capítulo 1 - Consultas a unWISE y AllWISE

## Consultar la tabla 'object' de unWISE

```python
query = """
        SELECT mag_w1_vg as w1, mag_w2_vg as w2, w1_w2_vg as w1_w2, 
               flux_w1/dflux_w1 as snr1, flux_w2/dflux_w2 as snr2, 
               ra, dec, glon, glat, spread_model_w1, dspread_model_w1 
        FROM unwise_dr1.object 
        WHERE (mag_w1_vg < 20 and mag_w2_vg < 20 and flux_w1/dflux_w1>5 and flux_w2/dflux_w2>5 and glat>50 and glat<60 and random_id>20 and random_id<21) 
        LIMIT 500000"""
print(query)

# Notas sobre la consulta:
# - Cortes de S/N en W1 y W2: flux_w1/dflux_w1>5 y flux_w2/dflux_w2>5
# - Cortes en latitud galáctica (glat) para restringir la misma región del cielo en unWISE y AllWISE
# - Corte en random_id para seleccionar aleatoriamente sobre el cielo (además de la condición glat)
# 
# - Para seleccionar por magnitudes en su lugar:
#        WHERE (mag_w1_vg < 20 and mag_w2_vg < 20 and glat>50 and glat<60 and random_id>20 and random_id<21)
#
```

```python
try:
    result = qc.query(sql=query,timeout=600) # por omisión, el resultado es una cadena con formato CSV
except Exception as e:
    print(e.message)
```

```python
# Convertir el resultado a un DataFrame de Pandas
df = convert(result,'pandas',na_values=('NaN','Infinity'))
```

```python
# Imprimir algunas filas del DataFrame
print(df[:5])
len(df)
```

## Consultar la tabla 'source' de AllWISE

```python
query = """
        SELECT w1mpro as w1, w2mpro as w2, w1snr, w2snr,
               ra, dec, glon, glat, random_id, ph_qual
        FROM allwise.source
        WHERE (w1snr>5 and w2snr>5 and glat>50 and glat<60 and random_id > 20 and random_id < 21)
        LIMIT 500000"""
print(query)
#
# w1(2)snr = relación señal-a-ruido en W1(2)
# ph_qual = marcador de calidad fotométrica
#
```

```python
try:
    result_all = qc.query(sql=query,timeout=600) # por omisión, el resultado es una cadena con formato CSV
except Exception as e:
    print(e.message)
```

```python
# Convertir el resultado a un DataFrame de Pandas
df_all = convert(result_all,'pandas',na_values=('NaN','Infinity'))
```

```python
# Imprimir algunas filas del DataFrame
print(df_all[:5])
len(df_all)
```

NOTA:
Ya podemos ver a partir de los resultados de la consulta anterior que hay más objetos en unWISE con S/N>5 tanto en W1 como en W2 (N=294,368) en comparación con AllWISE (N=124,764).

# Capítulo 2 - Distribuciones de Color y Magnitud

Como referencia, compararemos con las distribuciones del color *W1-W2* y la magnitud *W1* provenientes de la muestra cruzada AllWISE-SDSS DR10 del trabajo de [Kurcz et al. (2016)](https://ui.adsabs.harvard.edu/abs/2016A%26A...592A..25K). A continuación, generaremos distribuciones a partir de nuestros resultados de consulta.

## Color Distributions for AllWISE & unWISE DR1

```python
magthres = 16. #umbral de magnitud para investigar objetos brillantes (<16) vs. tenues (>16)
colthres = 0.1 #umbral de color en W1-W2 entre azul y rojo para las gráficas a continuación

# AllWISE (sufijo "_all")
faint_w1_all  = df_all['w1']>magthres
bright_w1_all = df_all['w1']<=magthres
w1_w2_all = df_all['w1'] - df_all['w2']
blue_all  = w1_w2_all<=colthres
red_all   = w1_w2_all>colthres

# unWISE
faint_w1  = df['w1']>magthres
bright_w1 = df['w1']<=magthres
blue = df['w1_w2']<=colthres
red  = df['w1_w2']>colthres
```

```python
# Distribución de color W1-W2
fig, ax = plt.subplots(figsize=(10,5))
ax.hist(w1_w2_all[bright_w1_all],bins=40,density=True,range=(-1,2),histtype='step',lw=3,color='Blue',alpha=0.3,label='AllWISE W1<16')
ax.hist(w1_w2_all[faint_w1_all],bins=40,density=True,range=(-1,2),histtype='step',lw=3,color='Red',alpha=0.4,label='AllWISE W1>16')
ax.set_xlabel('W1-W2 (Vega)',fontsize=15)
ax.set_ylabel('Histograma',fontsize=15)
ax.set_title('unWISE & AllWISE')
ax.hist(df['w1_w2'][bright_w1],bins=40,density=True,range=(-1,2),histtype='step',lw=3,color='Blue',alpha=1,label='unWISE W1<16')
ax.hist(df['w1_w2'][faint_w1],bins=40,density=True,range=(-1,2),histtype='step',lw=3,color='Red',alpha=1,label='unWISE W1>16')

ax.legend(loc='upper right',frameon=False,fontsize=14)
plt.show()
```

Podemos ver en la figura anterior que las distribuciones de color son similares entre AllWISE (histogramas de color claro) y unWISE (histogramas más darks) tanto para los objetos *brillantes* (`W1<16`, en azul) como para los *tenues* (`W1>16`, en rojo) en las tablas.

## Distribuciones de magnitud para AllWISE y unWISE DR1

```python
# Distribuciones de magnitud W1 (divididas entre W1-W2 azules y rojos) 
fig, ax = plt.subplots(figsize=(10,5.5))
ax.hist(df_all['w1'][blue_all],bins=40,density=False,range=(11,19),histtype='step',lw=3,color='Blue',alpha=0.4,label='AllWISE W1-W2<0.1')
ax.hist(df_all['w1'][red_all],bins=40,density=False,range=(11,19),histtype='step',lw=3,color='Red',alpha=0.4,label='AllWISE W1-W2>0.1')
ax.set_xlabel('W1 (Vega)',fontsize=15)
ax.set_ylabel('Histograma',fontsize=15)
ax.set_title('unWISE & AllWISE')
ax.hist(df['w1'][blue],bins=40,density=False,range=(11,19),histtype='step',lw=3,color='Blue',alpha=1,label='unWISE W1-W2<0.1')
ax.hist(df['w1'][red],bins=40,density=False,range=(11,19),histtype='step',lw=3,color='Red',alpha=1,label='unWISE W1-W2>0.1')

ax.legend(loc='upper left',frameon=False,fontsize=14)
plt.show()
```

La figura anterior muestra una comparación entre las distribuciones de magnitud de AllWISE (histogramas de colores claros) y unWISE (histogramas darks), divididas entre colores WISE azules (`W1-W2`<0.1, en azul) y colores rojos (`W1-W2`>0.1, en rojo). Los histogramas *no* están normalizados, de modo que también podamos comparar el número absoluto de objetos. Esto deja claro que la tabla de unWISE contiene muchos más objetos tenues, y que ambas tablas incluyen una mayor proporción de objetos rojos en comparación con los objetos azules.

## Diagrama Color-Magnitud: Gráficos de dispersión 

```python
# definir una figura de 2 paneles
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(14,6))

# primer panel
x = df['w2']
y = df['w1_w2']
ax1.scatter(x,y,0.2)
ax1.set_xlabel('W2')
ax1.set_ylabel('W1-W2')
ax1.set_xlim((19,4))
ax1.set_ylim((-1.5,2.5))
ax1.set_title('unWISE (S/N>5)')

# Saturación alrededor de magnitud 8 (puede dar colores/magnitudes inusuales)
ax1.axvline(x=8,color='Red')
ax1.axvline(x=17.5,color='Gray',lw=1,alpha=0.5)

# segundo panel
xx = df_all['w2']
yy = w1_w2_all
ax2.scatter(xx,yy,0.2)
ax2.set_xlabel('W2')
ax2.set_ylabel('W1-W2')
ax2.set_xlim((19,4))
ax2.set_ylim((-1.5,2.5))
ax2.set_title('AllWISE (S/N>5)')

# Saturación alrededor de magnitud 8 (puede dar colores/magnitudes inusuales)
ax2.axvline(x=8,color='Red')
ax2.axvline(x=17.5,color='Gray',lw=1,alpha=0.5)

plt.show()
```

La figura anterior muestra gráficos de dispersión de las distribuciones color-magnitud para unWISE (panel izquierdo) y AllWISE (panel derecho).  
La línea vertical gris está fija en `W2=17.5`, que está cerca del límite de detección de unWISE (con S/N>5).  
La línea roja vertical en `W2=8` marca la magnitud aproximada donde los objetos se saturan, lo que puede dar magnitudes y colores poco confiables más brillantes que este punto.  

Hay similitudes entre las dos distribuciones bi-variantes, como las características horizontales alrededor de `W1-W2=-0.1` y `0.1`, y un ensanchamiento significativo de la distribución de color para objetos más débiles que `W2=14`.
