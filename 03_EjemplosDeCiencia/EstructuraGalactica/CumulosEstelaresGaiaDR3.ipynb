# Configurando una consulta básica

La base de datos se llama **gaia_dr3** e incluye varias tablas. Exploramos las tablas disponibles a continuación.

```python
# Construir la consulta como cadena
query = "SELECT * FROM gaia_dr3.gaia_source LIMIT 1"

# Enviar la consulta
response = qc.query(sql=query, fmt='csv')

# Convertir la salida de cadena a un DataFrame de pandas
df_init = convert(response,'pandas')

# Para ver las columnas disponibles, incluir el siguiente comando
df_init.columns
```

```python
# Escribir una consulta que devuelva todas las estrellas alrededor de un objeto particular hasta un radio particular 

def input_query(object_name, radius_deg):
    # SkyCoord.from_name devuelve las coordenadas de un objeto dado su nombre como cadena
    c = SkyCoord.from_name(object_name)
    
    # Construir una consulta de Búsqueda Cónica (Cone Search)
    query = """SELECT * FROM gaia_dr3.gaia_source 
            WHERE (q3c_radial_query(ra, dec, {0}, {1}, %s)
            AND NOT parallax='Nan')""".format(c.ra.value,c.dec.value) % (radius_deg)
    
    # Enviar la consulta
    response = qc.query(sql=query, fmt='csv')
    
    # Convertir la salida a un DataFrame de pandas
    df = convert(response,'pandas')
    
    # Comprobar el número de objetos en tu consulta
    print("La consulta de %s contiene %s objetos." % (object_name, len(df)))
    
    # Devolver el DataFrame de salida
    return df
```

```python
# ¡Elige tu objeto favorito!

# Definir un objeto
object_queried = "NGC 6281"

# Definir un radio
radius = 0.5

# Ejecutar la función que devuelve un DataFrame consultado
df = input_query(object_queried, radius)
```

```python
# Definir variables a partir del DataFrame
# Hacerlo dentro de una función hace muy simple redefinirlas más adelante
def define_variables (df):
    ra, dec = df["ra"], df["dec"]
    ra_rad, dec_rad = np.radians(ra), np.radians(dec)
    pmra, pmdec = df["pmra"], df["pmdec"]
    parallax = df['parallax']
    g, g_bp = df["phot_g_mean_mag"], df["phot_g_mean_mag"]-df["phot_bp_mean_mag"]
    return ra, dec, ra_rad, dec_rad, pmra, pmdec, parallax, g, g_bp

# Ejecutar la función que define variables a partir del DataFrame
ra, dec, ra_rad, dec_rad, pmra, pmdec, parallax, g, g_bp = define_variables(df)
```

# Visualización estacionaria:

```python
# Escribir una función que genere gráficos espaciales y de color-magnitud para tu objeto muestreado
def panel_plot(ra, dec, pmra, pmdec, color, mag):
    fig = plt.figure(figsize=(26,8))
    
    # Un gráfico espacial simple
    ax1 = fig.add_subplot(131)
    ax1.scatter(ra, dec, marker='.', alpha= 1/np.power(len(df), 0.1))
    ax1.set_title("Gráfico espacial",fontsize=10)
    ax1.set_xlabel("AR",fontsize=10)
    ax1.set_ylabel("Declinación",fontsize=10)
    
    # Un gráfico de flechas (gráfico espacial con flechas que describen el movimiento propio de cada objeto)
    ax2 = fig.add_subplot(132)
    ax2.quiver(ra, dec, pmra, pmdec)
    ax2.set_title("Movimientos propios",fontsize=10)
    ax2.set_xlabel("AR",fontsize=10)
    ax2.set_ylabel("Declinación",fontsize=10)
    
    # Un diagrama color-magnitud que muestra la densidad de objetos
    ax3 = fig.add_subplot(133)
    im3 = ax3.hexbin(g_bp, g, bins='log', cmap=plt.cm.viridis, mincnt=1)
    ax3.set_ylim(max(g), min(g))
    ax3.set_title("Diagrama color-magnitud",fontsize=10)
    ax3.set_xlabel("Color Gaia G-BP",fontsize=10)
    ax3.set_ylabel("Magnitud Gaia G",fontsize=10)
    cb = plt.colorbar(im3,label='log(N)')
```

# Animación de paralaje:
Código basado en el de Eric Tollerud vía http://nbviewer.jupyter.org/gist/eteq/02a0065f15da3b3d8c2a9dea146a2a14/parallax_GaiaDR2.ipynb

Inspirado por James Davenport vía https://twitter.com/jradavenport/status/976205932274245632

```python
# Convertir variables a objeto SkyCoord e imponer límites en la salida

def convert_to_skycoord (g, ra_rad, dec_rad, parallax, dist1, dist2):
    # Definir variables como un arreglo de numpy y objeto SkyCoord
    # Nota: El objeto SkyCoord contiene datos de AR, Dec y paralaje
    mag = np.array(g)
    scnow = SkyCoord(np.array(ra_rad) * u.rad, np.array(dec_rad) * u.rad, 
                 distance=(np.array(parallax)*u.mas).to(u.pc, u.parallax()))

    # Crear una "máscara" con distintos umbrales de datos (aquí dependiendo de magnitud y distancia)
    msk = (mag < 16)  & (scnow.distance > dist1*u.pc) & (scnow.distance < dist2*u.pc)
    
    # Aplicar umbrales de datos a la magnitud dada y objetos SkyCoord
    mag = mag[msk]
    scnow = scnow[msk]
    
    # Imprimir el nuevo conteo de objetos después de aplicar los umbrales
    print('Después de aplicar los umbrales, esta consulta contiene %s objetos.' % (len(scnow)))
    
    # Retornar las variables definidas
    return mag, msk, scnow
    
mag, msk, scnow = convert_to_skycoord(g, ra_rad, dec_rad, parallax, 1e2, 9e2)
```

```python
# Hacer una versión especial del mapa de color viridis que vaya a un amarillo más oscuro al final. 
# Esto hace que sea más fácil ver puntos pequeños en un fondo blanco.
newvhsv = colors.rgb_to_hsv(cm.viridis.colors)
newvhsv[:,-1] = np.minimum(newvhsv[:,-1], .88)
darker_vidiris = colors.ListedColormap(colors.hsv_to_rgb(newvhsv))

# Definir una función para representar magnitud con tamaño del marcador y distancia con color del marcador
# Aquí el morado simbolizará los objetos más cercanos, y el amarillo simbolizará los más lejanos
def mag_to_size(mag):
    unity_rescaled_mag = (np.max(mag) - mag)/np.ptp(mag)
    return 40*unity_rescaled_mag**2+2

# Guardar estos argumentos de visualización como 'scatkwargs'
scatkwargs = dict(s=mag_to_size(mag), alpha=.9, c=scnow.distance, cmap=darker_vidiris)
```

```python
# Escribir una función para animar la paralaje de tu muestra

# La amplitud describe la distancia que moverán los marcadores al representar la paralaje.
amplitude = 0.6*u.lightyear
repr0 = scnow.barycentrictrueecliptic.cartesian

def animfunc(phase, eclcart_to_spherical, centerstr, ylims, ax):
    
    # Convertir de coordenadas cartesianas a esféricas
    drepr = CartesianRepresentation([np.cos(phase),np.sin(phase),0]*amplitude)
    rep = repr0 + drepr
    srep = eclcart_to_spherical(rep)
    
    # Crear y definir la estética de la gráfica
    ax.cla()
    ax.set_title(f'Center={centerstr}, Semieje Mayor de la Tierra={amplitude} \n')
    # ax.scatter(srep.lon.rad, 90-srep.lat.deg, **scatkwargs)
    ax.scatter(80.2+srep.lon.rad, 90-srep.lat.deg, **scatkwargs)
    ax.set_ylim(*ylims)
    ax.set_yticks(np.linspace(*ax.get_ylim(), 5)[1:])
    ax.set_yticklabels(['{}'.format(int(90-yl)) for yl in ax.get_yticks()])
    ax.grid(alpha=0.5)
```

```python
# Centrar la visualización en un objeto dado
def eclcart_to_spherical3(rep):
    icrsrep = SkyCoord(scnow.barycentrictrueecliptic.realize_frame(rep)).icrs.cartesian   
    rot1 = matrix_utilities.rotation_matrix(object_name.ra, 'z')
    rot2 = matrix_utilities.rotation_matrix(90*u.deg-object_name.dec, 'y')
    return icrsrep.transform(rot2 @ rot1).represent_as(SphericalRepresentation)
```
