# Leer la tabla de cúmulos desde la [base de datos gogreen_dr1](https://datalab.noirlab.edu/query.php?name=gogreen_dr1.clusters)

```python
cumulos = qc.query('select * from gogreen_dr1.clusters', fmt='pandas')
```

```python
cumulos.columns
```

```python
# Listar los nombres de los cúmulos, y la primera máscara de GOGREEN y de GCLASS para comprobar
# qué cúmulos no tienen ningún dato de GOGREEN (gogreen_m1 = NaN)
cumulos[['cluster','gogreen_m1','gclass_m1']]
```

```python
print('Número total de cúmulos de galaxias: ', len(cumulos))
cumulo = cumulos.cluster

cumulo
```

<a class="anchor" id="LF1"></a>
# Ubicación de los archivos en el servicio de archivos

```python
unaddir = 'gogreen_dr1://SPECTROSCOPY/OneD/'  # Espectros 1D
dosddir = 'gogreen_dr1://SPECTROSCOPY/TwoD/'  # Espectros 2D
imdir = 'gogreen_dr1://PHOTOMETRY/IMAGES/'    # Fotometría e imágenes

# Crear variables para nombres de archivos
unadarchivos = unaddir + cumulo + '_final.fits'
dosdarchivos = dosddir + cumulo + '_twod.fits.gz'
```

<a class="anchor" id="LF2"></a>
# Espectros 1D
Aquí se llama al servicio StoreClient como `sc` para recuperar los nombres de archivo. Para más usos e información, ver el cuaderno [How-to-use-the-StoreClient](https://github.com/astro-datalab/notebooks-latest/blob/master/04_HowTos/StoreClient/How_to_use_the_Data_Lab_StoreClient.ipynb).

```python
print(sc.ls(unaddir,format='long'))
```

```python
# Sólo está la versión del README en inglés en los directorios
print(wrapper.fill(text=sc.get(unaddir+'README')))
```

_(Traducción del README anterior)_  
Este directorio contiene los espectros finales 1D para cada sistema en GOGREEN y GCLASS. Estos han sido calibrados en flujo absoluto siempre que ha sido posible, usando las calibraciones v1.1 de Lyndsay.

### Nota:
De la lista anterior, hay 24 subcarpetas para 26 cúmulos. Esto es lo esperado porque dos pares de cúmulos están en el mismo campo y por lo tanto comparten un archivo (ver nota al pie de la Tabla 1 en el artículo de GOGREEN DR1):
- `SXDF49` y `SXDF87` comparten un único campo GMOS.  Los espectros de ambos están incluidos en los archivos fits de `SXDF49`.
- `SXDF76` y `SXDF76b` comparten un único campo GMOS.  Los espectros de ambos están incluidos en los archivos fits de `SXDF76`.

A continuación, verificaremos esto imprimiendo los nombres de los archivos de espectros 1D.

```python
for i in range(len(cumulos)):
     print("%-10s " % cumulo[i],end='')
     if sc.stat(unadarchivos[i]) != {}:
         print(True)
     else:
         print(False)
```

### Nota:
Como se esperaba, vemos que `SXDF76b` y `SXDF87` no aparecen listados con archivos de espectros 1D separados.

<a class="anchor" id="LF3"></a>
# Espectros 2D

```python
print(sc.ls(dosddir,format='long'))
```

```python
print(wrapper.fill(text=sc.get(dosddir+'README')))
```

_(Traducción del README anterior)_  
Cada archivo contiene espectros 2D de GOGREEN. El eje espacial está en píxeles, que corresponden a 0.16". No se ha aplicado ninguna calibración de flujo relativa o absoluta a estos espectros; las unidades de los píxeles están en cuentas del detector. Nótese que la dimensión de estos archivos MEF puede ser diferente de la de los archivos 1D correspondientes, porque estos últimos incluyen espectros de GCLASS.

```python
for i in range(len(cumulos)):
     print("%-10s " % cumulo[i],end='')
     if sc.stat(dosdarchivos[i]) != {}:
         print(True)
     else:
         print(False)
```

### Nota:
Los últimos cinco cúmulos SpARCS carecen de una máscara GOGREEN y, por lo tanto, no tienen carpeta de datos 2D (como se esperaba). Además, los dos pares de cúmulos que comparten un campo GMOS están agrupados como vimos para los espectros 1D arriba (`SXDF76b` junto con `SXDF76`; `SXDF87` junto con `SXDF49`).

# Imágenes

Ahora listamos las carpetas en el directorio de imágenes.

```python
print(sc.ls(imdir, format='long'))
```

### Nota:
Arriba vemos que la nomenclatura es diferente de la de los nombres de cúmulos de la tabla. Concretamente:
- `SPTxxxx` se nombran `SPTCL-xxxx`
- `SpARCSxxxx` se nombran `SpARCS-xxxx`

Hay 16 de 26 cúmulos con una carpeta de imágenes. En el caso del cúmulo `SpARCS1033`, la imagen en banda K no estaba disponible en el momento de la primera publicación. Sin embargo, estará disponible en el futuro como parte de GOGREEN DR2.

Adicionálmente, hay una carpeta `HST/` para imágenes del Telescopio Espacial Hubble. Examinemos su contenido a continuación.

```python
print(sc.ls(imdir+'HST/', format='long'))
```

### Nota:
Arriba vemos que la nomenclatura es diferente de la carpeta anterior, y también diferente de la de la tabla de cúmulos en un caso. Concretamente:
- `SPTxxxx` se nombran igual que en la tabla de cúmulos;
- `SpARCSxxxx` se nombran `SPARCSxxxx` (todas las letras en mayúsculas)

Hay 17 de 26 cúmulos con una carpeta de imágenes `HST/`.
