El método `login ()` devolverá un token válido o un mensaje de error.  Al llamar al método `isValidToken ()` es posible atrapar cualquier error de inicio de sesión en la tarea.  El uso de la función `` getPass () `` significa que las contraseñas no necesitan ser expuestas y los notebooks se pueden compartir más libremente.

<font color = 'rojo'> <h4> nota de seguridad </h4> </font>

<font color = 'verde'>
Una vez que haya iniciado sesión con éxito, se guarda un token de identidad en el servidor que ejecuta la aplicación (el servidor de notebook Jupyter en este caso) y se ofrecerá para inicios de sesión posteriores cuando no se proporcione contraseña.  Esto significa que los notebooks se pueden compartir para usar una cuenta válida sin compartir la contraseña en el notebook.  Puede usar el método `logrout ()` (ver más abajo) para eliminar este token del servidor manualmente, pero como precaución adicional, los tokens no se utilizarán en servidores públicos de notebook, y la contraseña se requerirá <b> siempre </b> al intentar iniciar sesión como algún usuario que otro usuario registrado en los servidores de notebooks auténticos.
</font>

#### Registro <i> out </i> del laboratorio de datos

El método `logrout ()` se utiliza para registrar al usuario fuera del laboratorio de datos y eliminar el token de identidad del servidor.  Estrictamente hablando, no se requiere que una aplicación cierre una vez que esté completa, los inicios de sesión posteriores devolverán el mismo token de identidad la próxima vez que inicie sesión siempre que sea válido (por ejemplo, no ha iniciado sesión en otro lugar)

```python
# Inicie sesión al usuario actual y elimine el token de identidad.
print ('Está conectado el token:', ac.isTokenLoggedIn(token))
res = ac.logout(token)
print('REGALO RESULTADO ="%s"'% res)
print('Es el token que se registra después del inicio de sesión ()::', ac.isTokenLoggedIn(token))
```

### Ejemplo: Guardar resultados en el almacenamiento virtual

Los detalles del uso del sistema de almacenamiento virtual se describen en otro lugar, pero en general el término *almacenamiento virtual *se utiliza para referirse a la base de datos privada de un usuario (su *myDB *) o un servicio de almacenamiento de archivos remoto que reside cerca de los servicios de datos.  * El almacenamiento virtual* se utiliza para permitir que se creen flujos de trabajo que minimicen la transferencia de datos.  

#### Guardar en el almacenamiento de archivos remoto

Guardar resultados en un archivo se puede hacer especificando el parámetro 'Out' en el formulario:

`Vos: // <FileName>`

El `<<silename>` debe ser único para cada consulta o de lo contrario se emitirá un error y los resultados no se guardarán.  Por lo tanto, un buen estilo de codificación atraparía posibles errores y lo manejaría en consecuencia, p.

```python
query = "select * from usno.b1 limit 20"
try:
    response = qc.query(token, adql=query, fmt='csv', out='vos://mags.csv')
except Exception as e:
    # Maneje cualquier error en la consulta.  Al ejecutar esta celda varias veces con el mismo
    # Archivo de salida, o mediante el uso de una instrucción SQL falsa, puede ver varios mensajes de error.
    print(e)
else:
    if response is not None: 
        print(response)           # Imprima la respuesta
    else:
        print("DE ACUERDO")
```

#### Guardar en la base de datos remota del usuario

Del mismo modo, los resultados se pueden guardar en la base de datos MYDB de un usuario especificando un nombre de tabla del formulario:

`mydb: // <tableName>`

```python
# Deja caer la tabla de prueba si ya existe
qc.mydb_drop('usno_test',token=token)

query = "select * from usno.b1 limit 1000"
try:
    response = qc.query (token, adql=query, fmt='csv', out='mydb://test1')
except Exception as e:
    # Manejar cualquier error en la consulta.  Ejecutando esta celda varias veces con la misma
    # Archivo de salida, o mediante el uso de una instrucción SQL falsa, puede ver varios mensajes de error.
    print(e)
else:
    if response is not None: 
        print(response)           # Imprima la respuesta
    else:
        print("DE ACUERDO")
```
