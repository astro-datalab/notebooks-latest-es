La mayoría de los notebooks comenzarán importando los paquetes necesarios. En este caso, querremos importar la interfaz ``authClient`` del paquete ``dl`` (es decir, Data Lab).

````python
from getpass import getpass
from dl import authClient as ac, queryClient as qc
````

El método `login()` devolverá un token válido o un mensaje de error.  Al llamar al método `isValidToken()` es posible detectar cualquier error de inicio de sesión en la tarea.  El uso de la función ``getpass()`` significa que no es necesario exponer las contraseñas y los notebooks se pueden compartir más libremente.

<font color='red'><h4>Nota de seguridad</h4></font>

<font color='green'>
Una vez que haya iniciado sesión correctamente, se guardará un token de identidad en el servidor que ejecuta la aplicación (el servidor de notebook Jupyter en este caso) y se ofrecerá para inicios de sesión posteriores cuando no se proporcione ninguna contraseña.  Esto significa que los notebooks se pueden compartir para usar una cuenta válida sin compartir la contraseña en el notebook. Puede utilizar el método `logout()` (ver más abajo) para eliminar este token del servidor manualmente, pero como precaución adicional, los tokens no se utilizarán en servidores de notebooks públicos y la contraseña <b>siempre</b> será solicitada al intentar iniciar sesión como algún usuario que no sea el usuario que inició sesión en servidores de notebooks autenticados.
</font>

#### Cerrar sesión <i>Cerrar</i> del Data Lab

El método `logout()` se utiliza para cerrar la sesión del usuario en Data Lab y eliminar el token de identidad del servidor.  Estrictamente hablando, no es necesario que una aplicación cierre sesión una vez completada; los inicios de sesión posteriores devolverán el mismo token de identidad la próxima vez que inicie sesión, siempre que siga siendo válido (por ejemplo, que no haya cerrado sesión en ningún otro lugar).

````python
# Cierre la sesión del usuario actual y elimine el token de identidad.
print ('¿El token ha iniciado sesión?', ac.isTokenLoggedIn(token))
res = ac.logout(token)
print('Resultado de cerrar sesión ="%s"'% res)
print('¿El token ha iniciado sesión después de cerrar sesión()?', ac.isTokenLoggedIn(token))
````

### Ejemplo: guardar resultados en almacenamiento virtual

Los detalles sobre el uso del sistema de almacenamiento virtual se describen en otra parte, pero en general el término *Almacenamiento virtual* se utiliza para referirse a la base de datos privada de un usuario (su *MyDB*) o a un servicio de almacenamiento de archivos remoto que reside cerca de los servicios de datos.  *Almacenamiento virtual* se utiliza para permitir la creación de flujos de trabajo que minimicen la transferencia de datos.

#### Guardar en el almacenamiento de archivos remoto

Se pueden guardar los resultados en un archivo especificando el parámetro 'out' de la forma:

`vos://<nombre de archivo>`

El `<nombre de archivo>` debe ser único para cada consulta o, de lo contrario, se emitirá un error y los resultados no se guardarán.  Por lo tanto, un buen estilo de codificación detectaría errores potenciales y los manejaría en consecuencia, por ejemplo:

````python
query = "select * from usno.b1 limit 20"
try:
    response = qc.query(token, adql=query, fmt='csv', out='vos://mags.csv')
except Exception as e:
    # Maneje cualquier error en la consulta.  Al ejecutar esta celda varias veces con el mismo
    # archivo de salida, o mediante el uso de una declaración SQL falsa, puede ver varios mensajes de error.
    print(e)
else:
    if response is not None: 
        print(response)           # imprimir la respuesta
    else:
        print("OK")
````

#### Guardar en la base de datos remota del usuario

De manera similar, los resultados se pueden guardar en la base de datos MyDB de un usuario especificando un nombre de tabla con el formato:

`mydb://<nombre de tabla>`

Al igual que con los nombres de archivos, `<nombre de tabla>` debe ser único.

````python
# Eliminar la tabla de prueba si ya existe
qc.mydb_drop('usno_test',token=token)

query = "select * from usno.b1 limit 1000"
try:
    response = qc.query (token, adql=query, fmt='csv', out='mydb://test1')
except Exception as e:
    # Maneje cualquier error en la consulta.  Al ejecutar esta celda varias veces con el mismo
    # archivo de salida, o mediante el uso de una declaración SQL falsa, puede ver varios mensajes de error.
    print(e)
else:
    if response is not None: 
        print(response)           # imprimir la respuesta
    else:
        print("OK")
````
