El kernel DRAGONS está disponible en el entorno Data Lab, lo que permite a los usuarios acceder a las rutinas sin depender de la instalación del software en sus máquinas locales.

Aquí presentamos un ejemplo de un notebook Jupyter de DRAGONS que funciona en el entorno de Data Lab para reducir completamente los datos de imágenes de banda Y de ejemplo de Gemini South Flamingos-2. Este notebook no presentará todos los detalles de las muchas opciones disponibles para ajustar u optimizar el proceso de reducción de datos de DRAGONS Flamingos-2; más bien, solo mostrará un ejemplo de una reducción estándar de un conjunto de datos de imágenes de Flamingos-2.

# Preparar el directorio de trabajo

Si tiene archivos intermedios que se crearon al ejecutar este código en el pasado, deberá eliminarlos de su directorio de trabajo. La siguiente celda define una función de limpieza que eliminará todos los archivos compatibles de su directorio de trabajo. Esta función se llamará nuevamente al final del tutorial, dejándolo solo con el producto final. De forma predeterminada, esta función eliminará todos los archivos en el directorio de trabajo. Si hay archivos que se han reducido previamente y que le gustaría conservar, configure `save_reduced=1` al llamar a la función.

```python
def clean_up(save_reduced=0):
    # ¿Ya existe el directorio de calibraciones?
    caldb_Exist = os.path.exists('./calibrations') 
    
    if caldb_Exist:
        shutil.rmtree('./calibrations', ignore_errors=True)

    # Elimine los archivos de registro y lista existentes.
    work_dir_path = os.getcwd()
    work_dir = os.listdir(work_dir_path)

    for item in work_dir:
        if item.endswith(".log") or item.endswith(".list"):
            os.remove(os.path.join(work_dir_path, item))
    
    # A continuación, eliminaremos todos los archivos de ajustes existentes, excepto los archivos previamente reducidos, dependiendo de cómo haya configurado save_reduced.
    if save_reduced:
        all_files_0 = glob.glob('*.fits')
        save = dataselect.select_data(all_files_0, [], ['PROCESSED'])
        
        for s in save:
            os.remove(os.path.join(work_dir_path,s))

    else:
        all_files_0 = glob.glob('*.fits')
        for a in all_files_0:
            os.remove(os.path.join(work_dir_path,a))
```

```python
clean_up(save_reduced=0)
```

# Crear listas de archivos

Este conjunto de datos contiene imágenes científicas y de calibración. Para algunos programas, podría haber diferentes objetivos observados y tiempos de exposición dependiendo de cómo organice sus datos sin procesar.

El proceso de reducción de datos de DRAGONS no organiza los datos por ti. Tu tienes que hacerlo. DRAGONS proporciona herramientas para ayudarte con eso.

El primer paso es crear listas que se utilizarán en el proceso de reducción de datos. Para eso utilizamos dataselect. Consulta la documentación de [dataselect](https://dragons.readthedocs.io/projects/recipe-system-users-manual/en/stable/supptools/dataselect.html?highlight=dataselect) para obtener detalles sobre su uso.

```python
dark_files_120s = dataselect.select_data(
    all_files,['F2', 'DARK', 'RAW'],[],
    dataselect.expr_parser('exposure_time==120'))

dark_files_2s = dataselect.select_data(
    all_files,['F2', 'DARK', 'RAW'],[],
    dataselect.expr_parser('exposure_time==2'))

list_of_flats_Y = dataselect.select_data(
     all_files,['FLAT'],[],
     dataselect.expr_parser('filter_name=="Y"'))

list_of_science_images = dataselect.select_data(
    all_files,['F2'],[],
    dataselect.expr_parser('(observation_class=="science" and filter_name=="Y")'))
```

<a class="anchor" id="Master_dark"></a>
# Crear un dark maestro

El dark maestro final tendrá el mismo nombre que el primer archivo dark, con el sufijo modificado _dark.fits.

```python
reduce_darks = Reduce()
reduce_darks.files.extend(dark_files_120s)
reduce_darks.runr()
```

<a class="anchor" id="BPM"></a>
# Crear una máscara de píxeles malos

La Bad Pixel Mask (BPM) se puede construir utilizando imágenes de flats con las lámparas encendidas y apagadas y un conjunto de archivos darks de corta exposición. Aquí, nuestros archivos darks más cortos tienen un tiempo de exposición de 2 segundos. Nuevamente usamos el comando `reduce` para producir las BPM.

Es importante tener en cuenta que la asociación de la biblioteca de recetas se basa en la naturaleza del primer archivo de la lista de entrada. Dado que la receta para hacer el BPM se encuentra en la biblioteca de recetas para flats, el primer elemento de la lista debe ser un flat.

Para Flamingos-2, la ubicación de la rueda de filtros es tal que la elección del filtro no interfiere con los resultados. Aquí usaremos flats de banda Y.

El BPM tendrá el nombre del primer archivo flat con el sufijo _bpm.fits.

```python
reduce_bpm = Reduce()
reduce_bpm.files.extend(list_of_flats_Y)
reduce_bpm.files.extend(dark_files_2s)
reduce_bpm.recipename = 'makeProcessedBPM'
reduce_bpm.runr()

bpm_filename = reduce_bpm.output_filenames[0]
```

# Agregue las máscaras de píxeles malos a la base de datos de calibración

```python
caldb.add_cal(bpm_filename)
```

<a class="anchor" id="Master_Flat"></a>
# Crear un campo flat maestro

El flat maestro de banda Y F2 se crea a partir de una serie de exposiciones con lámpara encendida y apagada. Todos deberían tener el mismo tiempo de exposición. Cada tipo se apila (promedia), luego la pila de lámpara apagada se resta de la pila de lámpara encendida y el resultado se normaliza.

El archivo flat maestro tendrá el nombre del primer archivo flat con el sufijo _flat.fits.

```python
reduce_flats = Reduce()
reduce_flats.files.extend(list_of_flats_Y)
reduce_flats.uparms = [('addDQ:user_bpm', bpm_filename)]
reduce_flats.runr()
```

<a class="anchor" id="Reduce_Science"></a>
# Reducir las imágenes científicas.
Este comando recupera el dark y el flat maestros y los aplica a los datos científicos. Para la substracción del cielo, el software analiza la secuencia para establecer si se trata de una secuencia de oscilación (dither) en el objetivo o de compensación del cielo y procede en consecuencia. Finalmente, las imágenes restadas del cielo se alinean y se apilan juntas. Las fuentes en las imágenes se utilizan para la alineación.

El archivo del producto final tendrá el sufijo _stack.fits.

Las unidades de la pila de salida están en electrones (palabra clave del encabezado BUNIT=electrons). La pila de salida se almacena en un archivo FITS de múltiples extensiones (MEF). La señal científica está en la extensión "SCI", la varianza está en la extensión "VAR" y el plano de calidad de datos (máscara) está en la extensión "DQ".

Se crearán varios archivos al ejecutar esta celda. Se crearán dos archivos nuevos para cada imagen científica. Ambos archivos contendrán el nombre del archivo, el primero contendrá el sufijo _flatCorrected.fits y el segundo _skyCorrected.fits. Además de estos archivos, la imagen apilada final compartirá el nombre del primer archivo científico con el sufijo _stack.fits.

```python
reduce_target = Reduce()
reduce_target.files.extend(list_of_science_images)
reduce_target.uparms = [('addDQ:user_bpm', bpm_filename)]
reduce_target.runr()
```
