# Resumen
DRAGONS es una plataforma de reducción de datos astronómicos basada en Python escrita por el Departamento de soporte al usuario de Gemini Science. Actualmente se puede utilizar para reducir datos de imágenes de los instrumentos Gemini GMOS, NIRI, Flamingos 2, GSAOI y GNIRS, así como datos espectroscópicos tomados con GHOST y GMOS en modo de rendija larga. Enlazado <a href="https://dragons.readthedocs.io/en/v4.0.0/">aquí</a> hay una lista general de guías, manuales y tutoriales sobre el uso de DRAGONS.

El kernel DRAGONS está disponible en el entorno Data Lab, lo que permite a los usuarios acceder a las rutinas sin depender de la instalación del software en sus máquinas locales.

Aquí, presentamos un ejemplo de un notebook Jupyter de DRAGONS que funciona en el entorno de Data Lab para reducir por completo los datos de imágenes de banda G de Gemini South GMOS. Este notebook no presentará todos los detalles de las muchas opciones disponibles para ajustar u optimizar el proceso de reducción de datos de DRAGONS GMOS; en cambio, solo mostrará un ejemplo de una reducción estándar de un conjunto de datos de imágenes GMOS.

Los datos utilizados en este ejemplo de notebook son imágenes de banda G GMOS del archivo Gemini de la galaxia NGC 5018 del programa Gemini South "<a href="https://archive.gemini.edu/programinfo/GS-2018A-Q-207">The Evolutionary History of NGC 5018</a>", IP: L. Sesto, ID de programa GS-2018A-Q-207. La imagen científica reducida final combina cinco fotogramas científicos de 460 segundos cada uno que se difuminaron entre cada exposición.

# Preparar el directorio de trabajo

Si tiene archivos intermedios que se crearon al ejecutar este código en el pasado, deberá eliminarlos de su directorio de trabajo. La siguiente celda define una función de limpieza que eliminará todos los archivos fits de su directorio de trabajo. Esta función se llamará nuevamente al final del tutorial, dejándolo solo con el producto final. De forma predeterminada, esta función eliminará todos los archivos en el directorio de trabajo. Si hay archivos que se han reducido previamente y que le gustaría conservar, configure `save_reduced=1` al llamar a la función.

```python
def clean_up(save_reduced=0):
    # ¿Ya existe el directorio de calibraciones?
    caldb_Exist = os.path.exists('./calibrations') 
    
    if caldb_Exist:
        shutil.rmtree('./calibrations', ignore_errors=True)

    # Elimine los archivos de registro y lista existentes.
    work_dir_path = os.getcwd()
    work_dir = os.listdir(work_dir_path)

    for item in work_dir:
        if item.endswith(".log") or item.endswith(".list"):
            os.remove(os.path.join(work_dir_path, item))
    
    # A continuación, eliminaremos todos los archivos fits existentes, excepto los archivos previamente reducidos, dependiendo de cómo haya configurado save_reduced.
    if save_reduced:
        all_files_0 = glob.glob('*.fits')
        save = dataselect.select_data(all_files_0, [], ['PROCESSED'])
        
        for s in save:
            os.remove(os.path.join(work_dir_path,s))

    else:
        all_files_0 = glob.glob('*.fits')
        for a in all_files_0:
            os.remove(os.path.join(work_dir_path,a))
```

```python
clean_up(save_reduced=0)
```

# Configurando el registrador de DRAGONS

DRAGONS viene con un administrador de calibraciones local que utiliza las mismas reglas de asociación de calibración que el Archivo del Observatorio Gemini. Esto permite que `reduce` realice solicitudes a una base de datos local liviana para hacer coincidir las calibraciones procesadas cuando sea necesario para reducir un conjunto de datos.

Esto le indica al sistema dónde colocar la base de datos de calibración. Esta base de datos realizará un seguimiento de las calibraciones procesadas que le vamos a enviar.

```python
logutils.config(file_name='gmos_data_reduction.log')
caldb = cal_service.set_local_database()
caldb.init("w")
```

```python
all_files = glob.glob('S2018*[0-9].fits')
all_files.sort()
```

<a class="anchor" id="File_Lists"></a>
# Crear listas de archivos

El primer paso es crear listas de archivos de entrada. La herramienta "dataselect" ayuda con eso. Utiliza etiquetas y "descriptores" de Astrodata para seleccionar los archivos y enviar los nombres de los archivos a un archivo de texto que luego se puede alimentar para "reducir". (Consulta el [Manual del usuario de Astrodata](https://dragons.readthedocs.io/_/downloads/astrodata-user-manual/en/v2.1.0/pdf/) para obtener información sobre Astrodata).

```python
list_biases  = dataselect.select_data(all_files,['BIAS'],[])
list_flats   = dataselect.select_data(all_files,['FLAT'],[],
                                      dataselect.expr_parser('filter_name=="g"'))
list_science = dataselect.select_data(all_files,[],['CAL'],
                                      dataselect.expr_parser('(observation_class=="science" and filter_name=="g")'))
```

<a class="anchor" id="Master_Bias"></a>
# Crea un bias maestro
Comenzamos la reducción de datos creando un bias maestro para los datos científicos. Se puede crear y agregar a la base de datos de calibración usando los siguientes comandos. Cuando la celda termine de ejecutarse, el archivo de bias maestro se guardará con el nombre del primer bias y el sufijo _bias.fits.

```python
reduce_bias = Reduce()
reduce_bias.files.extend(list_biases)
reduce_bias.runr()
```

<a class="anchor" id="Master_Flat"></a>
# Crear un campo flat maestro

Las imágenes flats de crepúsculo se utilizan para producir un flat maestro de imágenes y el resultado se agrega a la base de datos de calibración.

El flat maestro tendrá el nombre del primer archivo flat de Crepúsculo con el sufijo _flat.fits.

```python
reduce_flats = Reduce()
reduce_flats.files.extend(list_flats)
reduce_flats.runr()
```

<a class="anchor" id="Master_Fringe"></a>
# Crear imagen fringe maestra

El conjunto de datos utilizado en este tutorial no requiere una imagen fringe maestra. Sin embargo, la siguiente celda demuestra cómo crear una para que esté completa.

El margen maestro tendrá el nombre del primer archivo científico con el sufijo _fringe.fits.

```python
reduce_fringe = Reduce()
reduce_fringe.files.extend(list_science)
reduce_fringe.recipename = 'makeProcessedFringe'
reduce_fringe.runr()
```

<a class="anchor" id="Reduce_Science"></a>
# Reducir las imágenes científicas

Una vez que nuestros archivos de calibración se procesan y agregan a la base de datos, podemos ejecutar `reduce` en nuestros datos científicos.

Este comando generará archivos con corrección de bias y flats y los apilará. Si se necesita un fringe, este comando aplicará la corrección. La imagen apilada tendrá el sufijo _stack.

Las unidades de la pila de salida están en electrones (palabra clave del encabezado BUNIT=electrons). La pila de salida se almacena en un archivo FITS de múltiples extensiones (MEF). La señal científica está en la extensión "SCI", la varianza está en la extensión "VAR" y el flat de calidad de datos (máscara) está en la extensión "DQ".

Cada imagen científica reducida tendrá el nombre original con el sufijo _image.fits.

```python
reduce_science = Reduce()
reduce_science.files.extend(list_science)
reduce_science.runr()
```

<a class="anchor" id="Display_Image"></a>
# Mostrar la imagen apilada

```python
image_file = "S20180419S0098_image.fits"
hdu_list = fits.open(image_file)
wcs = WCS(hdu_list[1].header)
hdu_list.info()
```

```python
image_data = fits.getdata(image_file, ext=1)
print(image_data.shape)
```

```python
plt.figure(figsize = (15,15))
plt.subplot(projection=wcs)
plt.imshow(image_data,cmap='gray',norm=LogNorm(vmin=3000, vmax=60000),origin='lower')
plt.xlabel('Ascensión Recta [hh:mm:ss]',fontsize=14,fontweight='bold')
plt.ylabel('Declinación [grado]',fontsize=14,fontweight='bold')
plt.xlim(900,2500)
plt.ylim(1600,600)
plt.show()
```
