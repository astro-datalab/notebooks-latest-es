# Preparar el directorio de trabajo

Si tiene archivos intermedios que se crearon al ejecutar este código en el pasado, deberá eliminarlos de su directorio de trabajo. La siguiente celda define una función de limpieza que eliminará todos los archivos compatibles de su directorio de trabajo. Esta función se llamará nuevamente al final del tutorial, dejándolo solo con el producto final. De forma predeterminada, esta función eliminará todos los archivos en el directorio de trabajo. Si hay archivos que se han reducido previamente y que le gustaría conservar, configure `save_reduced=1` al llamar a la función.

```python
def clean_up(save_reduced=0):
    # ¿Ya existe el directorio de calibraciones?
    caldb_Exist = os.path.exists('./calibrations') 
    
    if caldb_Exist:
        shutil.rmtree('./calibrations', ignore_errors=True)

    # Elimine los archivos de registro y lista existentes.
    work_dir_path = os.getcwd()
    work_dir = os.listdir(work_dir_path)

    for item in work_dir:
        if item.endswith(".log") or item.endswith(".list"):
            os.remove(os.path.join(work_dir_path, item))
    
    # A continuación, eliminaremos todos los archivos de ajustes existentes, excepto los archivos previamente reducidos, dependiendo de cómo haya configurado save_reduced.
    if save_reduced:
        all_files_0 = glob.glob('*.fits')
        save = dataselect.select_data(all_files_0, [], ['PROCESSED'])
        
        for s in save:
            os.remove(os.path.join(work_dir_path,s))

    else:
        all_files_0 = glob.glob('*.fits')
        for a in all_files_0:
            os.remove(os.path.join(work_dir_path,a))
```

```python
clean_up(save_reduced=0)
```

Esta es una observación de imágenes de ojo de cerradura de adquisición GNIRS de una fuente puntual. La secuencia de observación utiliza tramado en el objetivo. Se utilizarán observaciones difuminadas cercanas en el tiempo para la resta del cielo de cada cuadro.

Una nota sobre la búsqueda de los oscuros en GOA: GNIRS no es un generador de imágenes y las imágenes a través del ojo de la cerradura se realizan solo en circunstancias extremas, por lo que el archivo no tiene reglas de asociación de calibración para los oscuros. # Crear listas de archivos

Este conjunto de datos contiene marcos científicos y de calibración. Algunos programas pueden tener diferentes objetivos observados y tiempos de exposición dependiendo de cómo organice sus datos sin procesar.

El proceso de reducción de datos de DRAGONS no organiza los datos por usted. Tienes que hacerlo. DRAGONS proporciona herramientas para ayudarte con eso.

El primer paso es crear listas que se utilizarán en el proceso de reducción de datos. Para eso utilizamos dataselect. Consulte la documentación de [dataselect](https://dragons.readthedocs.io/projects/recipe-system-users-manual/en/stable/supptools/dataselect.html?highlight=dataselect) para obtener detalles sobre su uso.

```python
darks60 = dataselect.select_data(
    all_files,
    ['DARK'],
    [],
    dataselect.expr_parser('exposure_time==60')
)

flats = dataselect.select_data(all_files, ['FLAT'])

target = dataselect.select_data(
    all_files,
    [],
    [],
    dataselect.expr_parser('object=="GRB120116A"')
)
```

<a class="anchor" id="Master_dark"></a>
# Crear maestro oscuro

Primero creamos la oscuridad maestra para el objetivo científico. El nombre del maestro oscuro de salida, N20120102S0538_dark.fits, se escribe en la pantalla al final del proceso.

```python
reduce_darks = Reduce()
reduce_darks.files.extend(darks60)
reduce_darks.runr()
```

<a class="anchor" id="Master_Flat"></a>
# Crear campo plano maestro

Un piso maestro GNIRS se crea a partir de una serie de exposiciones con lámparas encendidas y apagadas. Cada sabor se apila y luego la pila sin lámpara se resta de la pila con lámpara encendida.

El piso maestro se guardará con el sufijo _flat.fits.

```python
reduce_flats = Reduce()
reduce_flats.files.extend(flats)
reduce_flats.runr()
```

<a class="anchor" id="Reduce_Science"></a>
# Reducir las imágenes científicas

El objetivo científico es una fuente puntual. La secuencia oscila en el objetivo, moviendo la fuente a través de la delgada abertura del ojo de la cerradura. Los fotogramas del cielo para cada imagen científica serán los fotogramas difuminados adyacentes obtenidos dentro de un límite de tiempo determinado. El valor predeterminado para las imágenes de ojo de cerradura GNIRS es "dentro de 600 segundos".

Tanto el maestro oscuro como el maestro plano se encuentran en nuestra base de datos de calibración local. El administrador de calibración recuperaría automáticamente cualquier otro instrumento de las instalaciones de Gemini. Sin embargo, GNIRS no es un generador de imágenes y el ojo de cerradura normalmente se usa solo para adquisición, por lo que no existen reglas de asociación de calibración entre las imágenes de ojo de cerradura y las imágenes oscuras de GNIRS.

Las unidades de la pila de salida están en electrones (palabra clave del encabezado BUNIT=electrones). La pila de salida se almacena en un archivo FITS (MEF) de múltiples extensiones. La señal científica está en la extensión "SCI", la varianza está en la extensión "VAR" y el plano de calidad de datos (máscara) está en la extensión "DQ".

La imagen científica reducida final y la imagen apilada tendrán el mismo nombre que el primer archivo científico con los sufijos _image.fits y _stack.fits respectivamente.

NOTA: Después de ejecutar esta celda, es posible que reciba una ADVERTENCIA sobre muy pocos objetos correlacionados. Esto no afectará el producto final, por lo que puedes ignorarlo.

```python
reduce_target = Reduce()
reduce_target.files.extend(target)
reduce_target.uparms = [('darkCorrect:dark', 'N20120102S0538_dark.fits')]
reduce_target.runr()
```

# Mostrar la imagen apilada

```python
image_file = "N20120117S0014_image.fits"
hdu_list = fits.open(image_file)
wcs = WCS(hdu_list[1].header)
hdu_list.info()
```

```python
image_data = fits.getdata(image_file, ext=1)
print(image_data.shape)
```

```python
plt.figure(figsize = (10,10))
plt.subplot(projection=wcs)
plt.imshow(image_data,cmap='bone',norm=Normalize(vmin=1, vmax=500),origin='lower')
plt.xlabel('Ascensión Recta [hh:mm:ss]',fontsize=14,fontweight='atrevido')
plt.ylabel('Declinación [grado]',fontsize=14,fontweight='atrevido')
plt.show()
```
