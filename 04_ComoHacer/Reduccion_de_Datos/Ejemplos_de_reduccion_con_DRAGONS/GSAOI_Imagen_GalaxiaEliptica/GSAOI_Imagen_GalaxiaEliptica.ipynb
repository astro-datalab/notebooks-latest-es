# Preparar el directorio de trabajo

Si tiene archivos intermedios que se crearon al ejecutar este código en el pasado, deberá eliminarlos de su directorio de trabajo. La siguiente celda define una función de limpieza que eliminará todos los archivos compatibles de su directorio de trabajo. Esta función se llamará nuevamente al final del tutorial, dejándolo solo con el producto final. De forma predeterminada, esta función eliminará todos los archivos en el directorio de trabajo. Si hay archivos que se han reducido previamente y que le gustaría conservar, configure `save_reduced=1` al llamar a la función.

```python
def clean_up(save_reduced=0):
    # ¿Ya existe el directorio de calibraciones?
    caldb_Exist = os.path.exists('./calibrations') 
    
    if caldb_Exist:
        shutil.rmtree('./calibrations', ignore_errors=True)

    # Elimine los archivos de registro y lista existentes.
    work_dir_path = os.getcwd()
    work_dir = os.listdir(work_dir_path)

    for item in work_dir:
        if item.endswith(".log") or item.endswith(".list"):
            os.remove(os.path.join(work_dir_path, item))
    
    # A continuación, eliminaremos todos los archivos de ajustes existentes, excepto los archivos previamente reducidos, dependiendo de cómo haya configurado save_reduced.
    if save_reduced:
        all_files_0 = glob.glob('*.fits')
        save = dataselect.select_data(all_files_0, [], ['PROCESSED'])
        
        for s in save:
            os.remove(os.path.join(work_dir_path,s))

    else:
        all_files_0 = glob.glob('*.fits')
        for a in all_files_0:
            os.remove(os.path.join(work_dir_path,a))
```

```python
clean_up(save_reduced=0)
```

# Crear listas de archivos

El primer paso es crear listas de archivos de entrada. La herramienta "dataselect" ayuda con eso. Utiliza etiquetas y "descriptores" de Astrodata para seleccionar los archivos y enviar los nombres de los archivos a un archivo de texto que luego se puede alimentar para "reducir". <a class="anchor" id="Master_Flat"></a>
# Crear campo plano maestro

El piso maestro GSAOI Kshort se crea a partir de una serie de exposiciones de cúpula con y sin lámpara. Todos deberían tener el mismo tiempo de exposición. Cada sabor se apila (promedia), luego la pila de lámpara apagada se resta de la pila de lámpara encendida y el resultado se normaliza.

El piso maestro se guardará con el sufijo _flat.fits.

```python
reduce_flats = Reduce()
reduce_flats.files.extend(list_of_flats_Ks)
reduce_flats.runr()
```

<a class="anchor" id="Standard_Star"></a>
# Reducir estrella estándar

La estrella estándar se reduce esencialmente como el objetivo científico (siguiente sección). El campo plano procesado que agregamos a la base de datos de calibración local se recuperará automáticamente. Además, en este caso, la estrella estándar se obtuvo utilizando ROI (regiones de interés), que no coinciden con el campo plano. El software reconocerá que el campo plano es válido y lo recortará para que coincida con las ROI.

Esta celda creará tres archivos para cada uno de los oscuros científicos. Contendrán los sufijos _countsScaled.fits, _flatCorrected.fits y sourcesDetected.fits.

```python
reduce_std = Reduce()
reduce_std.files.extend(list_of_std_stars)
reduce_std.runr()
```

<a class="anchor" id="Reduce_Science"></a>
# Reducir las imágenes científicas

Esta es una observación de una galaxia desplazada hacia el cielo. Necesitamos desactivar la compensación aditiva del cielo porque el objetivo llena el campo de visión y no representa un fondo de cielo razonable. Si la compensación no se desactiva en este caso particular, se producirá una sustracción excesiva del marco del cielo.

**Nota**:
A diferencia de otros instrumentos de infrarrojo cercano, el parámetro aditivo offset_sky se utiliza de forma predeterminada para ajustar el fondo del marco del cielo para GSAOI en lugar del parámetro multiplicativo scale_sky. Se descubrió que funciona mejor cuando el fondo del cielo por píxel es muy bajo. Esto es común debido al corto tiempo de exposición necesario para evitar saturar las estrellas y a la pequeña escala de píxeles. Se anima al lector a experimentar con scale_sky si offset_sky no parece conducir a una resta de cielo óptima.

(Recuerde que ambos parámetros normalmente deben estar desactivados cuando se extiende la fuente).

El marco del cielo proviene de observaciones del cielo fuera del objetivo. Alimentamos el proceso con todos los marcos dentro y fuera del objetivo. El software dividirá adecuadamente los objetivos dentro y fuera del objetivo utilizando la información de los encabezados.

Este comando generará archivos con corrección plana y con resta de cielo, pero no los apilará. Puedes diferenciar cada archivo por su sufijo (_flatCorrected o _skyCorrected). Los archivos de destino son aquellos a los que se les ha restado el cielo (_skyCorrected). Finalmente, la imagen apilada se guardará con el sufijo _image.fits.

NOTA: Después de ejecutar esta celda, es posible que reciba ADVERTENCIAS de que no hay REFCAT disponible para algunos archivos. Estas ADVERTENCIAS no afectarán el producto final y pueden ignorarse.

```python
reduce_target = Reduce()
reduce_target.files.extend(list_of_science_images)
reduce_target.uparms['skyCorrect:offset_sky'] = False
reduce_target.runr()
```

<a class="anchor" id="Display_Image"></a>
# Mostrar la imagen apilada

Las unidades de la pila de salida están en electrones (palabra clave del encabezado BUNIT=electrones). La pila de salida se almacena en un archivo FITS (MEF) de múltiples extensiones. La señal científica está en la extensión "SCI", la varianza está en la extensión "VAR" y el plano de calidad de datos (máscara) está en la extensión "DQ".

```python
image_file = 'S20170505S0095_image.fits'
hdu_list = fits.open(image_file)
wcs = WCS(hdu_list[1].header)
hdu_list.info()
```

```python
image_data = fits.getdata(image_file, ext=1)
print(image_data.shape)
```

```python
plt.figure(figsize = (15,15))
plt.subplot(projection=wcs)
plt.imshow(image_data,cmap='bone',norm=Normalize(vmin=1, vmax=1000),origin='lower')
plt.xlabel('Ascensión Recta [hh:mm:ss]',fontsize=14,fontweight='atrevido')
plt.ylabel('Declinación [grado]',fontsize=14,fontweight='atrevido')
plt.show()
```
