## Buscar espectros en SPARCL
Usando el método `client.find()`, obtenemos los siguientes campos del conjunto de datos DESI DR1 usando SPARCL:
- `sparcl_id`: Identificador universalmente único para el espectro en SPARCL
- `ra`: Ascensión Recta en grados
- `dec`: Declinación en grados
- `spectype`: Tipo espectral del objeto (STAR, GALAXY o QSO)
- `subtipo`: subtipo espectral
- `specid`: identificador de espectro específico del conjunto de datos (puede no ser único)
- `redshift`: desplazamiento al rojo medido
- `redshift_err`: Incertidumbre sobre el corrimiento al rojo medido

Y aplique las siguientes restricciones:
- `2.0 <desplazamiento al rojo <5.0`
- `redshift_warning = 0` (máscara de bits de advertencia de desplazamiento al rojo medida por Redrock)
- `specprimary = 1` (indicador booleano (Verdadero/Falso) para el espectro primario combinado para un objeto objetivo determinado)
- `tipo de especificación = QSO`
- `lanzamiento_datos = DESI DR1`

```python
outfields = ['sparcl_id', 'ra', 'dec', 'spectype', 'subtype', 'specid',
             'redshift', 'redshift_err',]
constraints = {'redshift': [2.0, 5.0],
               'redshift_warning': [0],
               'specprimary': [1],
               'spectype': ['QSO'],
               'data_release': ['DESI-DR1'],
               }
```

```python
found = client.find(outfields=outfields,
                    constraints=constraints,
                    limit=50)
```

## # DESINAMBRE

La siguiente función está copiada de [`desiutil.names`](https://github.com/desihub/desiutil/blob/main/py/desiutil/names.py). `DESINAME` es una conveniente etiqueta RA basada en Dec que es consistente con los estándares IAU y NED.

Aunque a partir de DESI DR1, `DESINAME` está disponible en la base de datos SPARCL, no es un campo `CORE`, por lo que `client.find()` no lo devolverá. Sin embargo, necesitamos una etiqueta para que estos espectros pasen a `Specviz`, *antes* de recuperar los espectros. Así que calcularemos `DESINAME` como solución alternativa.

```python
def radec_to_desiname(target_ra, target_dec):
"""Convertir la ascensión recta y la declinación de un objetivo DESI
    en el correspondiente "DESINAME" para referencia en publicaciones.
    La longitud de target_ra y target_dec debe ser la misma si se proporciona un
    matriz o lista. Tenga en cuenta que estos nombres no son únicos para aproximadamente
    uno por ciento de los objetivos DESI, por lo que también se incluye TARGETID en
    Las publicaciones son muy recomendables por su singularidad.

    Parámetros
    ----------
    target_ra: matriz de :clase:`float64`
        Ascensión recta en grados del objeto(s) objetivo. Puede ser flotante, doble,
         o matriz/lista de flotadores o dobles
    target_dec: matriz de :clase:`float64`
        Declinación en grados del objeto(s) objetivo. Puede ser flotante, doble,
         o matriz/lista de flotadores o dobles

    Devoluciones
    -------
    matriz de :clase:`str`
        Los nombres DESI se refieren a los RA y DEC de destino de entrada. La matriz es
        la misma longitud que las matrices de entrada.

    """
    # Convertir a una matriz numpy en caso de que las entradas sean escalares o listas
    target_ra, target_dec = np.atleast_1d(target_ra), np.atleast_1d(target_dec)

    # Número de decimales en la convención de nomenclatura final
    precision = 4

    # Truncar decimales a la precisión dada
    ratrunc = np.trunc((10 ** precision) * target_ra).astype(int).astype(str)
    dectrunc = np.trunc((10 ** precision) * target_dec).astype(int).astype(str)

    # Recorra los valores de entrada y cree DESINAME como: DESI JXXX.XXXX+/-YY.YYYY
    # Aquí J se refiere a J2000, lo cual no es estrictamente correcto pero es el más cercano.
    # Término compatible con la IAU
    desinames = []
    for ra, dec in zip(ratrunc, dectrunc):
        desiname = 'DESI J' + ra[:-precision].zfill(3) + '.' + ra[-precision:]
        # Los números positivos necesitan un "+" explícito, mientras que los números negativos
        # ya tiene un "-".
        # zfill funciona correctamente con '-' pero lo cuenta en número de caracteres
        # así que necesito uno más
        if dec.startswith('-'):
            desiname += dec[:-precision].zfill(3) + '.' + dec[-precision:]
        else:
            desiname += '+' + dec[:-precision].zfill(2) + '.' + dec[-precision:]
        desinames.append(desiname)

    return np.array(desinames)
```

Cree `DESINAME` a partir del RA y Dec de cada registro:

```python
for record in found.records:
    if record['_dr'] == 'DESI-DR1':
        record['DESINAME'] = radec_to_desiname(record['ra'], record['dec']).tolist()[0]
metadata = sorted(found.records, key=lambda x: x['DESINAME'])
ids = [m['sparcl_id'] for m in metadata]
```

## Recuperar espectros de SPARCL y preparar datos de graficado
Aquí definimos una función que:
1. Utiliza el método `client.retrieve()` de SPARCL para recuperar espectros para la lista de ID de DESI DR1.
2. Crea un objeto `specutils.Spectrum1D` a partir de los datos recuperados.
3. Carga el espectro, los metadatos y el modelo.
4. Almacena en caché el objeto `Spectrum1D` convertido.

```python
def on_selected(b):
"""Función de devolución de llamada para permitir el cambio entre diferentes espectros DESI.

    Parámetros
    ----------
    b: objeto
        Un objeto de devolución de llamada proporcionado por Specviz.
    """
    specunit = u.Unit('10-17 erg cm-2 s-1 AA-1')
    key = entries[b['new']][0]
    result = client.retrieve(uuid_list=[ids[b['new']]],
                             include=['specid',
                                      'survey',
                                      'ra',
                                      'dec',
                                      'redshift',
                                      'redshift_err',
                                      'flux',
                                      'ivar',
                                      'wavelength',
                                      'mask',
                                      'model'])
    if key not in spectra:
        #
        # Observe cómo los datos y el modelo se preparan como dos objetos separados.
        #
        spectra[key] = (Spectrum1D(spectral_axis=result.records[0]['wavelength']*u.AA,
                                   flux=result.records[0]['flux']*specunit,
                                   uncertainty=InverseVariance(result.records[0]['ivar']*(specunit**-2)).represent_as(StdDevUncertainty),
                                   mask=result.records[0]['mask'],
                                   redshift=result.records[0]['redshift']),
                        Spectrum1D(spectral_axis=result.records[0]['wavelength']*u.AA,
                                   flux=result.records[0]['model']*specunit))
    if 'old' in b:
        #
        # Elimina el espectro anterior.
        #
        data0 = specviz.app.data_collection[0]
        data1 = specviz.app.data_collection[1]
        specviz.app.data_collection.remove(data0)
        specviz.app.data_collection.remove(data1)
    #
    # Cargue el espectro.
    #
    specviz.load_data(spectra[key][0], data_label=key + ' Data')
    opt = specviz.plugins['Plot Options']
    opt.layer = key + ' Data'
    opt.line_color.value = '# 000000' # Negro
    opt.line_as_steps = True
    opt.uncertainty_visible = True
    #
    # Agregar metadatos.
    #
    meta0 = specviz.app.data_collection[0].meta
    meta0['RA'] = result.records[0]['ra']
    meta0['Dec'] = result.records[0]['dec']
    meta0['redshift'] = result.records[0]['redshift']
    meta0['redshift uncertainty'] = result.records[0]['redshift_err']
    #
    # Cargue el modelo.
    #
    specviz.load_data(spectra[key][1], data_label=key + ' Model')
    opt = specviz.plugins['Plot Options']
    opt.layer = key + ' Model'
    opt.line_color.value = '# FF0000' # Rojo
    opt.line_as_steps = True
    #
    # Esta es una solución para un error en el que los metadatos no se muestran inicialmente.
    #
    m = specviz.plugins['Metadata']
    m.dataset = key + ' Model'
    m.dataset = key + ' Data'
```

## Graficar espectros usando Jdaviz
Ejecute la celda a continuación para abrir el gráfico interactivo en un nuevo panel a la derecha. Puede encontrar información sobre las opciones de la barra de herramientas en la [Guía del usuario de Jdaviz](https://jdaviz.readthedocs.io/en/latest/specviz/displaying.html).

Esta celda también mostrará un menú desplegable que controla el espectro que se mostrará en el gráfico interactivo.

```python
#
# Configure Specviz y el menú desplegable.
#
spectra = dict()  # Esto almacenará en caché los espectros recuperados.
entries = [(m['DESINAME'], i) for i, m in enumerate(metadata)]
select = widgets.Dropdown(options=entries, value=0, description='Spectrum:',)
output = widgets.Output()
specviz = Specviz()
#
# Active la descarga si este es el primer espectro.
#
on_selected({'new': 0})
#
# Conecte la acción de la función al menú desplegable.
#
select.observe(on_selected, names='value')
#
# Abra la pantalla en una pestaña separada.
#
specviz.show('sidecar:split-bottom')
#
# Muestra el menú desplegable.
#
display(select, output)
```
