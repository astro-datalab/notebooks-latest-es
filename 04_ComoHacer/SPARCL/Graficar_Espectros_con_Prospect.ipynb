## DESIDR1

### Encuentra espectros DESI

SPARCL proporciona acceso a espectros DESI que han sido apilados por HEALPixel. Esto corresponde a las entradas en la tabla `desi_dr1.zpix` en Astro Data Lab (consulte la información de la columna [aquí](https://datalab.noirlab.edu/query.php?name=desi_dr1.zpix)).

```python
q = """
SELECT z.targetid, z.chi2, z.z, z.zerr, z.zwarn, z.spectype, z.subtype,
       z.coadd_numexp, z.coadd_exptime, z.healpix, z.deltachi2
FROM desi_dr1.zpix AS z
WHERE z.zcat_primary
    AND z.survey = 'main'
    AND z.program = 'dark'
    AND z.spectype = 'GALAXY'
    AND z.z BETWEEN 0.5 AND 0.9
    AND z.zwarn = 0
ORDER BY z.targetid
LIMIT 50
"""
desi_ids = qc.query(sql=q, fmt='table')
desi_ids
```

**PR**: ¿Realmente tenemos 50 espectros únicos?

```python
assert (np.unique(desi_ids['targetid']) == desi_ids['targetid']).all()
```

### Recuperar espectros DESI

Con el conjunto de `targetid` obtenido anteriormente, podemos recuperar directamente los espectros DESI.

```python
include = client.get_all_fields(dataset_list=['DESI-DR1'])
desi_spectra = client.retrieve_by_specid(specid_list=desi_ids['targetid'].value.tolist(),
                                         include=include,
                                         dataset_list=['DESI-DR1'])
desi_spectra.info
```

**PR**: ¿Realmente encontramos todos los espectros esperados?

```python
targetids_sorted = sorted(desi_spectra.records, key=lambda x: x.targetid)
assert (np.array([r.targetid for r in targetids_sorted]) == desi_ids['targetid']).all()
assert all([r.survey == 'main' for r in desi_spectra.records])
assert all([r.program == 'dark' for r in desi_spectra.records])
```

**PR**: ¿Todos los registros de espectros tienen la misma solución de longitud de onda?

```python
assert all([(r.wavelength == desi_spectra.records[0].wavelength).all() for r in desi_spectra.records])
```

**PR**: ¿Alguno de los espectros está completamente enmascarado?

```python
desi_fully_masked = np.zeros((len(desi_spectra.records), ), dtype=bool)

for k in range(len(desi_spectra.records)):
    assert np.isfinite(desi_spectra.records[k].flux).all()
    assert np.isfinite(desi_spectra.records[k].ivar).all()
    if (desi_spectra.records[k].mask > 0).all():
        print(f"ADVERTENCIA: ¡El registro de espectro {k:d} está completamente enmascarado!")
        desi_fully_masked[k] = True
```

#### Ver el contenido del primer registro recuperado

```python
desi_spectra.records[0]
```

### Organizar metadatos

El cliente potencial necesita varias entradas:

1. Un objeto que contiene espectros.  En este caso usaremos [`prospect.specutils.Spectra`](https://desi-prospect.readthedocs.io/en/latest/api.html#prospect.specutils.Spectra), que hereda de [`SpectrumList`](https://specutils.readthedocs.io/en/stable/api/specutils.SpectrumList.html#specutils.SpectrumList), y en realidad es solo una objeto [`Spectrum1D`](https://specutils.readthedocs.io/en/stable/api/specutils.Spectrum1D.html#specutils.Spectrum1D) subyacente.
* El objeto contiene el flujo, la longitud de onda y la incertidumbre habituales.
* Además se necesita una tabla "fibermap" (mapa de fibras). Esta debería ser una `Tabla` de Astropy con las columnas esperadas.
2. Un catálogo de corrimientos al rojo. Esta debería ser una `Tabla` de Astropy con las columnas esperadas.
3. Un espectro modelo.  En realidad, el modelo lo proporciona SPARCL, pero debemos ingresarlo por separado.

#### Objeto de espectro

Primero ensamblamos los componentes del objeto del espectro.

```python
flux = np.zeros((len(desi_spectra.records), desi_spectra.records[0].flux.shape[0]),
                dtype=desi_spectra.records[0].flux.dtype)
uncertainty = np.zeros((len(desi_spectra.records), desi_spectra.records[0].ivar.shape[0]),
                       dtype=desi_spectra.records[0].ivar.dtype)
mask = np.zeros((len(desi_spectra.records), desi_spectra.records[0].mask.shape[0]),
                dtype=desi_spectra.records[0].mask.dtype)

meta = {'sparcl_id': list(), 'data_release': list()}
sparcl_id = list()
data_release = list()

for k in range(len(desi_spectra.records)):
    flux[k, :] = desi_spectra.records[k].flux
    uncertainty[k, :] = desi_spectra.records[k].ivar
    mask[k, :] = desi_spectra.records[k].mask
    meta['sparcl_id'].append(desi_spectra.records[k].sparcl_id)
    meta['data_release'].append(desi_spectra.records[k].data_release)
```

Y la tabla "fibermap". Empezaremos con cantidades fotométricas.

```python
columns = ('targetid', 'ra', 'dec', 'ref_epoch', 'pmra', 'pmdec', 'ebv',
           'flux_g', 'flux_r', 'flux_z', 'flux_w1', 'flux_w2')
q = """
SELECT {0}
FROM desi_dr1.photometry
WHERE targetid IN ({1})
ORDER BY targetid
""".format(', '.join(columns), ', '.join([str(t) for t in desi_ids['targetid'].value.tolist()]))

fibermap = qc.query(sql=q, fmt='table')
```

```python
for col in fibermap.colnames:
    if col == 'ra' or col == 'dec':
        fibermap.rename_column(col, 'TARGET_' + col.upper())
    else:
        fibermap.rename_column(col, col.upper())
```

**PR**: ¿Encontramos fotometría para cada `targetid`?

```python
assert (fibermap['TARGETID'] == desi_ids['targetid']).all()
```

A continuación agregamos máscaras de bits de orientación.

```python
columns = ('targetid', 'desi_target', 'bgs_target', 'mws_target', 'scnd_target')
q = """
SELECT DISTINCT {0}
FROM desi_dr1.target
WHERE targetid IN ({1})
ORDER BY targetid
""".format(', '.join(columns), ', '.join([str(t) for t in fibermap['TARGETID'].value.tolist()]))

targeting = qc.query(sql=q, fmt='table')
```

```python
for col in targeting.colnames:
    targeting.rename_column(col, col.upper())
```

**PR**: ¿Encontramos seleccionamiento para cada `targetid`?

```python
assert (targeting['TARGETID'] == fibermap['TARGETID']).all()
```

Agregue columnas en `fibermap`.

```python
for col in ('DESI_TARGET', 'BGS_TARGET', 'MWS_TARGET', 'SCND_TARGET'):
    fibermap.add_column(targeting[col])
fibermap
```

Finalmente ensambla el objeto.

```python
desi_prospect = Spectra(bands=['coadd'],
                        wave={'coadd': desi_spectra.records[0].wavelength},
                        flux={'coadd': flux[~desi_fully_masked, :]},
                        ivar={'coadd': uncertainty[~desi_fully_masked, :]},
                        mask={'coadd': mask[~desi_fully_masked, :]},
                        fibermap=fibermap[~desi_fully_masked],
                        meta={'coadd': meta})
```

#### Catálogo de corrimiento al rojo

Podemos reutilizar la consulta inicial anterior; fue construido deliberadamente.

```python
desi_zcatalog = desi_ids.copy()
for col in desi_zcatalog.colnames:
    if col == 'healpix':
        desi_zcatalog.rename_column(col, 'HPXPIXEL')
    else:
        desi_zcatalog.rename_column(col, col.upper())
desi_zcatalog
```

#### Espectros modelo

Prospect espera un modelo en forma de tupla que contenga longitud de onda y flujo. Dado que SPARCL proporciona el modelo, esto es fácil.  Hay otras formas de especificar el modelo, pero requieren más acceso a la pila de software DESI *y* a* los *archivos* de datos.

```python
model_flux = np.zeros((len(desi_spectra.records), desi_spectra.records[0].model.shape[0]),
                      dtype=desi_spectra.records[0].model.dtype)

for k in range(len(desi_spectra.records)):
        model_flux[k, :] = desi_spectra.records[k].model

desi_model = (desi_spectra.records[0].wavelength, model_flux[~desi_fully_masked, :])
```

### Iniciar prospecto

Con todo ensamblado, la interfaz para prospectar es solo una llamada.  Tenga en cuenta que estamos configurando `with_vi_widgets=True` para describir completamente la visualización del cliente potencial. Otros ejemplos de visualización a continuación tendrán `with_vi_widgets=False`.

Más detalles sobre la visualización de prospectos:

* La ventana grande muestra un espectro con el modelo de mejor ajuste superpuesto.
* Al escanear el gráfico con el mouse se muestra un zoom de resolución completa de esa porción del espectro a la derecha, lo cual es particularmente útil para escanear líneas de emisión estrechas.
* Pruebe las diversas herramientas en la parte superior para realizar panorámicas, hacer zoom y seleccionar. Cuando se selecciona la herramienta de zoom con rueda (predeterminada), el uso de la rueda de desplazamiento acercará y alejará el zoom. Apunte el mouse sobre el eje xoy antes de desplazarse para hacer zoom solo en ese eje.
* Haga clic en una de las entradas de la leyenda en la parte inferior derecha para activar o desactivar ese elemento.
* El siguiente ejemplo incluye herramientas de inspección visual (VI), en un cuadro sombreado en la parte inferior izquierda. Esto permite a los individuos calificar los espectros en preguntas como "¿es exacto el corrimiento al rojo?" o "¿es el objeto realmente una galaxia en lugar de una estrella?". Las pautas sobre el uso de las herramientas VI se encuentran justo debajo de las herramientas mismas.
* Inmediatamente encima de las herramientas VI hay flechas y un control deslizante para mostrar otros espectros.
* A la derecha de las herramientas VI hay metadatos sobre el espectro, intercalados con herramientas para suavizar el espectro, cambiar manualmente el corrimiento al rojo del modelo de mejor ajuste y alternar la visualización de los marcadores de líneas de emisión y absorción.

```python
plotspectra(desi_prospect,
            zcatalog=desi_zcatalog[~desi_fully_masked],
            redrock_cat=None,
            notebook=True,
            with_thumb_tab=False,
            with_vi_widgets=True,
            with_coaddcam=False,
            mask_type='DESI_TARGET',
            model_from_zcat=False,
            model=desi_model)
```

### Organizar metadatos

3. Un espectro modelo.  En realidad, el modelo lo proporciona SPARCL, pero debemos ingresarlo por separado.
