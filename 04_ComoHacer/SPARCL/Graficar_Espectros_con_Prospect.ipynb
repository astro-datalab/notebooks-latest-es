## Resumen y objetivos

[prospect](https://github.com/desihub/prospect/) es una herramienta interactiva de visualización de espectros creada por la [colaboración DESI](https://www.desi.lbl.gov) para visualizar y caracterizar espectros de DESI. También ha sido adaptada para mostrar datos de otros proyectos, en particular del [Sloan Digital Sky Survey (SDSS)](https://www.sdss.org).

*Proyecto*: Obtener espectros de DESI DR1, SDSS DR17 y (e)BOSS DR17 utilizando el [servicio de espectros SPARCL de NOIRLab](https://astrosparcl.datalab.noirlab.edu), convertir los datos a [objetos de specutils](https://specutils.readthedocs.io/en/stable/) según sea necesario y usar [prospect](https://github.com/desihub/prospect/) para visualizar los datos.

*Punto clave*: En ningún momento de este notebook se abrirán archivos de datos, excepto en una sección “Aside”.

*Punto clave*: prospect permite que múltiples visualizaciones de espectros independientes coexistan dentro del mismo notebook.

Este notebook es una adaptación del notebook [Prospect + specutils + SPARCL](https://github.com/desihub/prospect/blob/main/doc/nb/Prospect_spectrum_service.ipynb). La figura inferior muestra una porción de la visualización de prospect con un ejemplo de un espectro DESI: el espectro observado se grafica en rojo (suavizado mediante un pequeño kernel gaussiano con $\sigma=1$), el mejor modelo ajustado por el pipeline en negro y la incertidumbre en verde (almacenada como la varianza inversa `ivar`). Los espectros de DESI y SDSS incluyen además un arreglo `mask` que indica si algunos píxeles son problemáticos (en cuyo caso `mask!=0`), y los espectros de SDSS incluyen un espectro de cielo (`sky`), que no se muestra ni se utiliza en este notebook. La herramienta prospect también incluye una imagen en color en la ubicación de la fibra, y la herramienta interactiva permite al usuario hacer clic en la imagen en miniatura para abrir el [Legacy Surveys Viewer](https://www.legacysurvey.org/viewer) en esa posición del cielo en una pestaña separada del navegador.

## DESIDR1

### Encuentra espectros DESI

SPARCL proporciona acceso a espectros DESI que han sido apilados por HEALPixel. Esto corresponde a las entradas en la tabla `desi_dr1.zpix` en Astro Data Lab (consulte la información de la columna [aquí](https://datalab.noirlab.edu/query.php?name=desi_dr1.zpix)).

```python
q = """
SELECT z.targetid, z.chi2, z.z, z.zerr, z.zwarn, z.spectype, z.subtype,
       z.coadd_numexp, z.coadd_exptime, z.healpix, z.deltachi2
FROM desi_dr1.zpix AS z
WHERE z.zcat_primary
    AND z.survey = 'main'
    AND z.program = 'dark'
    AND z.spectype = 'GALAXY'
    AND z.z BETWEEN 0.5 AND 0.9
    AND z.zwarn = 0
ORDER BY z.targetid
LIMIT 50
"""
desi_ids = qc.query(sql=q, fmt='table')
desi_ids
```

**PR**: ¿Realmente tenemos 50 espectros únicos?

`assert` garantiza que la expresión que se evalúa sea ``True`` y se utiliza en muchas pruebas estándar en Python.

```python
assert (np.unique(desi_ids['targetid']) == desi_ids['targetid']).all()
```

### Recuperar espectros DESI

Con el conjunto de `targetid` obtenido anteriormente, podemos recuperar directamente los espectros DESI.

```python
include = client.get_all_fields(dataset_list=['DESI-DR1'])
desi_spectra = client.retrieve_by_specid(specid_list=desi_ids['targetid'].value.tolist(),
                                         include=include,
                                         dataset_list=['DESI-DR1'])
desi_spectra.info
```

**PR**: ¿Realmente encontramos todos los espectros esperados?

Efecto secundario: esto también ordena los registros devueltos para que coincidan con el orden de las entradas.

```python
desi_records = sorted(desi_spectra.records, key=lambda x: x.targetid)
assert (np.array([r.targetid for r in desi_records]) == desi_ids['targetid']).all()
assert all([r.survey == 'main' for r in desi_records])
assert all([r.program == 'dark' for r in desi_records])
```

**PR**: ¿Todos los registros de espectros tienen la misma solución de longitud de onda?

```python
assert all([(r.wavelength == desi_records[0].wavelength).all() for r in desi_records])
```

**PR**: ¿Alguno de los espectros está completamente enmascarado?

```python
desi_fully_masked = np.zeros((len(desi_records), ), dtype=bool)

for k in range(len(desi_records)):
    assert np.isfinite(desi_records[k].flux).all()
    assert np.isfinite(desi_records[k].ivar).all()
    if (desi_records[k].mask > 0).all():
        print(f"ADVERTENCIA: ¡El registro de espectro {k:d} está completamente enmascarado!")
        desi_fully_masked[k] = True
```

### Organizar metadatos

El cliente potencial necesita varias entradas:

1. Un objeto que contiene espectros.  En este caso usaremos [`prospect.specutils.Spectra`](https://desi-prospect.readthedocs.io/en/latest/api.html#prospect.specutils.Spectra), que hereda de [`SpectrumList`](https://specutils.readthedocs.io/en/stable/api/specutils.SpectrumList.html#specutils.SpectrumList), y en realidad es solo una objeto [`Spectrum1D`](https://specutils.readthedocs.io/en/stable/api/specutils.Spectrum1D.html#specutils.Spectrum1D) subyacente.
* El objeto contiene el flujo, la longitud de onda y la incertidumbre habituales.
* Además se necesita una tabla "fibermap" (mapa de fibras). Esta debería ser una `Tabla` de Astropy con las columnas esperadas.
2. Un catálogo de corrimientos al rojo. Esta debería ser una `Tabla` de Astropy con las columnas esperadas.
3. Un espectro modelo.  En realidad, el modelo lo proporciona SPARCL, pero debemos ingresarlo por separado.

#### Objeto de espectro

Primero ensamblamos los componentes del objeto del espectro.

```python
flux = np.zeros((len(desi_records), desi_records[0].flux.shape[0]),
                dtype=desi_records[0].flux.dtype)
uncertainty = np.zeros((len(desi_records), desi_records[0].ivar.shape[0]),
                       dtype=desi_records[0].ivar.dtype)
mask = np.zeros((len(desi_records), desi_records[0].mask.shape[0]),
                dtype=desi_records[0].mask.dtype)

meta = {'sparcl_id': list(), 'data_release': list()}
sparcl_id = list()
data_release = list()

for k in range(len(desi_records)):
    flux[k, :] = desi_records[k].flux
    uncertainty[k, :] = desi_records[k].ivar
    mask[k, :] = desi_records[k].mask
    meta['sparcl_id'].append(desi_records[k].sparcl_id)
    meta['data_release'].append(desi_records[k].data_release)
```

Y la tabla "fibermap". Empezaremos con cantidades fotométricas.

```python
columns = ('targetid', 'ra', 'dec', 'ref_epoch', 'pmra', 'pmdec', 'ebv',
           'flux_g', 'flux_r', 'flux_z', 'flux_w1', 'flux_w2')
q = """
SELECT {0}
FROM desi_dr1.photometry
WHERE targetid IN ({1})
ORDER BY targetid
""".format(', '.join(columns), ', '.join([str(t) for t in desi_ids['targetid'].value.tolist()]))

fibermap = qc.query(sql=q, fmt='table')
```

```python
for col in fibermap.colnames:
    if col == 'ra' or col == 'dec':
        fibermap.rename_column(col, 'TARGET_' + col.upper())
    else:
        fibermap.rename_column(col, col.upper())
```

**PR**: ¿Encontramos fotometría para cada `targetid`?

```python
assert (fibermap['TARGETID'] == desi_ids['targetid']).all()
```

A continuación agregamos máscaras de bits de orientación.

```python
columns = ('targetid', 'desi_target', 'bgs_target', 'mws_target', 'scnd_target')
q = """
SELECT DISTINCT {0}
FROM desi_dr1.target
WHERE targetid IN ({1})
ORDER BY targetid
""".format(', '.join(columns), ', '.join([str(t) for t in fibermap['TARGETID'].value.tolist()]))

targeting = qc.query(sql=q, fmt='table')
```

```python
for col in targeting.colnames:
    targeting.rename_column(col, col.upper())
```

**PR**: ¿Encontramos seleccionamiento para cada `targetid`?

```python
assert (targeting['TARGETID'] == fibermap['TARGETID']).all()
```

**PR**: ¿El orden del fibermap coincide con los records de SPARCL?

```python
assert (np.array([r.targetid for r in desi_records]) == fibermap['TARGETID']).all()
```

Agregue columnas en `fibermap`.

```python
for col in ('DESI_TARGET', 'BGS_TARGET', 'MWS_TARGET', 'SCND_TARGET'):
    fibermap.add_column(targeting[col])
fibermap
```

Finalmente ensambla el objeto.

```python
desi_prospect = Spectra(bands=['coadd'],
                        wave={'coadd': desi_records[0].wavelength},
                        flux={'coadd': flux[~desi_fully_masked, :]},
                        ivar={'coadd': uncertainty[~desi_fully_masked, :]},
                        mask={'coadd': mask[~desi_fully_masked, :]},
                        fibermap=fibermap[~desi_fully_masked],
                        meta={'coadd': meta})
```

#### Catálogo de corrimiento al rojo

Podemos reutilizar la consulta inicial anterior; fue construido deliberadamente.

```python
desi_zcatalog = desi_ids.copy()

# prospect espera que los nombres de las columnas estén en letras mayúsculas y que la columna HEALPIX
# se llame en su lugar HPXPIXEL, por lo que hacemos estos cambios a continuación
for col in desi_zcatalog.colnames:
    if col == 'healpix':
        desi_zcatalog.rename_column(col, 'HPXPIXEL')
    else:
        desi_zcatalog.rename_column(col, col.upper())
desi_zcatalog
```

#### Espectros modelo

Prospect espera un modelo en forma de tupla que contenga longitud de onda y flujo. Dado que SPARCL proporciona el modelo, esto es fácil.  Hay otras formas de especificar el modelo, pero requieren más acceso a la pila de software DESI *y* a* los *archivos* de datos.

```python
model_flux = np.zeros((len(desi_records), desi_records[0].model.shape[0]),
                      dtype=desi_records[0].model.dtype)

for k in range(len(desi_records)):
        model_flux[k, :] = desi_records[k].model

desi_model = (desi_records[0].wavelength, model_flux[~desi_fully_masked, :])
```

### Aparte: espectros modelo en DESI DR1

### Organizar metadatos

El cliente potencial necesita varias entradas:

1. Un objeto que contiene espectros.  En este caso usaremos un objeto [`Spectrum1D`](https://specutils.readthedocs.io/en/stable/api/specutils.Spectrum1D.html#specutils.Spectrum1D).
    * El objeto contiene el flujo, la longitud de onda y la incertidumbre habituales.
    * Además, se necesita una tabla "plugmap". Esta debería ser una `Tabla` de Astropy con las columnas esperadas.
2. Un catálogo de corrimientos al rojo. Esta debería ser una `Tabla` de Astropy con las columnas esperadas.
3. Un espectro modelo.  En realidad, el modelo lo proporciona SPARCL, pero debemos ingresarlo por separado.

#### Objeto de espectro

Primero ensamblamos los componentes del objeto del espectro.

```python
flux = np.zeros((len(sdss_spectra.records), sdss_spectra.records[0].flux.shape[0]),
                dtype=sdss_spectra.records[0].flux.dtype)
uncertainty = np.zeros((len(sdss_spectra.records), sdss_spectra.records[0].ivar.shape[0]),
                       dtype=sdss_spectra.records[0].ivar.dtype)
mask = np.zeros((len(sdss_spectra.records), sdss_spectra.records[0].mask.shape[0]),
                dtype=sdss_spectra.records[0].mask.dtype)

meta = {'sparcl_id': list(), 'data_release': list()}
sparcl_id = list()
data_release = list()

for k in range(len(sdss_spectra.records)):
    flux[k, :] = sdss_spectra.records[k].flux
    uncertainty[k, :] = sdss_spectra.records[k].ivar
    mask[k, :] = sdss_spectra.records[k].mask
    meta['sparcl_id'].append(sdss_spectra.records[k].sparcl_id)
    meta['data_release'].append(sdss_spectra.records[k].data_release)
```

Y la tabla "plugmap". Empezaremos con cantidades fotométricas.

```python
columns = ('objid', 'ra', 'dec', 'u', 'g', 'r', 'i', 'z')
q = """
SELECT {0}
FROM sdss_dr16.photoplate
WHERE objid IN ({1})
ORDER BY objid
""".format(', '.join(columns), ', '.join([str(t) for t in sdss_ids['bestobjid'].value.tolist()]))

plugmap = qc.query(sql=q, fmt='table')
```

```python
for col in plugmap.colnames:
    plugmap.rename_column(col, col.upper())
mag = np.zeros((len(plugmap), 5), dtype=plugmap['G'].value.dtype)
for k, band in enumerate('UGRIZ'):
    mag[:, k] = plugmap[band].value
plugmap.add_column(mag, name='MAG')
plugmap.add_column(sdss_ids['primtarget'], name='PRIMTARGET')
plugmap.add_column(sdss_ids['sectarget'], name='SECTARGET')
meta['plugmap'] = plugmap[~sdss_fully_masked]
```

**PR**: ¿Encontramos fotometría para cada "bestobjid"?

```python
assert (plugmap['OBJID'] == sorted(sdss_ids['bestobjid'])).all()
```

Finalmente ensambla el objeto.

```python
sdss_prospect = Spectrum1D(flux=flux[~sdss_fully_masked, :] * u.Unit('1e-17 erg / (Angstrom cm2 s)'),
                           spectral_axis=sdss_spectra.records[0].wavelength * u.Unit('Angstrom'),
                           uncertainty=InverseVariance(uncertainty[~sdss_fully_masked, :]),
                           mask=mask[~sdss_fully_masked, :] != 0,
                           meta=meta)
```

#### Catálogo de desplazamiento al rojo

Podemos reutilizar la consulta inicial anterior; fue construido deliberadamente.

```python
sdss_zcatalog = sdss_ids.copy()
for col in sdss_zcatalog.colnames:
    if col == 'zerr':
        sdss_zcatalog.rename_column(col, 'Z_ERR')
    else:
        sdss_zcatalog.rename_column(col, col.upper())
sdss_zcatalog
```

#### Espectros modelo

Prospect espera un modelo en forma de tupla que contenga longitud de onda y flujo. Dado que SPARCL proporciona el modelo, esto es fácil.  Hay otras formas de especificar el modelo, pero requieren más acceso a los *archivos* SDSS.

```python
model_flux = np.zeros((len(sdss_spectra.records), sdss_spectra.records[0].model.shape[0]),
                      dtype=sdss_spectra.records[0].model.dtype)

for k in range(len(sdss_spectra.records)):
        model_flux[k, :] = sdss_spectra.records[k].model

sdss_model = (sdss_spectra.records[0].wavelength, model_flux[~sdss_fully_masked, :])
```

### Iniciar prospecto

Con todo ensamblado, la interfaz para prospectar es solo una llamada.

```python
plotspectra(sdss_prospect,
            zcatalog=sdss_zcatalog[~sdss_fully_masked],
            redrock_cat=None,
            notebook=True,
            with_thumb_tab=False,
            with_vi_widgets=False,
            with_coaddcam=False,
            mask_type='PRIMTARGET',
            model_from_zcat=False,
            model=sdss_model)
```
