# Escribir una consulta con una declaración de unión
Para extraer solo las columnas relevantes relacionadas con nuestra pregunta científica de múltiples tablas de datos, podemos escribir una consulta que use una declaración de unión. Hay 4 tipos principales de declaraciones de unión que podríamos usar, y cuál que decidimos elegir depende de cómo queremos extraer la información. 
1. ** (Inner) Únete **: Devuelve filas que tienen valores coincidentes en ambas tablas
2. ** Izquierda (externa) Únete **: Devuelve todas las filas de la mesa izquierda y las filas emparejadas desde la mesa derecha
3. ** Derecha (externa) Únete **: Devuelve todas las filas de la tabla derecha y las filas emparejadas desde la mesa izquierda
4. ** Full (externo) unirse **: Devuelve todas las filas cuando hay una coincidencia en la mesa izquierda o derecha

Tómese un momento para mirar sobre la figura a continuación que describe los diversos tipos de declaración de unión.  
Nota: La unión predeterminada es una `unión interna '.

<img src = 'join.png'> </img>

### `Únete lateral`
En las consultas de CrossMatch de vecinos más cercanos, usamos `unión lateral`, que es como un bucle SQL foreach que iterará sobre cada fila en un conjunto de resultados y evaluará una subconsulta utilizando esa fila como parámetro.

# Ejemplos de uso de una declaración de unión en una consulta con catálogos LS y SDSS
## Ejemplo 1: una sola unión
Primero examinaremos el desplazamiento al rojo espectroscópico de los objetos que se encuentran tanto en el catálogo SDSS DR17 como en el catálogo LS DR9 escribiendo una consulta con una sola declaración de unión entre su tabla pre-cross y la tabla SDSS DR17. Las dos tablas CrossMatch relacionadas con estos dos catálogos son:

`ls_dr9.x1p5__tractor__sdss_dr17__specobj`  
`sdss_dr17.x1p5__specobj__ls_dr9__tractor`

La elección de cuáles de estas dos tablas CrossMatch para usar deben basarse en la pregunta científica que se plantea. Por ejemplo, la pregunta * '¿Cómo cambia la estructura de una galaxia con el desplazamiento al rojo?' * * Depende de los valores de desplazamiento rojo obtenidos de SDSS DR17, por lo que debemos usar la tabla CrossMatch que tiene SDSS DR17 como la primera tabla. Por lo tanto, la información relevante que queremos seleccionar de nuestras dos tablas de interés para este ejemplo son:

1. "X" = `SDSS_DR17.X1P5__SPECOBJ__LS_DR9__tractor`
    - ** RA1 ** (RA del objeto SDSS)
    - ** Dic1 ** (dec del objeto SDSS)
2. "S" = `SDSS_DR17.specobj`
    - ** Z ** (Redshift)

### Escriba la consulta de la declaración de unión única
Ahora que sabemos de qué queremos y de dónde lo queremos, escribamos la consulta e imprima los resultados en la pantalla. Aquí usamos una declaración de unión: buscará en la tabla SDSS DR17 `Specobj` para filas que tienen el mismo valor de identificación de SDSS (` specobjid`) como en la tabla previa a la interrupción (`ID1`) y recuperan las columnas deseadas de la tabla SD17` Specobj` pre-ross dentro de la región especificada de RA y dec.

```python
query_single = ("""
SELECT 
    X.ra1 AS ra_sdss, X.dec1 AS dec_sdss,
    S.z
FROM
    sdss_dr17.x1p5__specobj__ls_dr9__tractor AS X 
JOIN
    sdss_dr17.specobj AS S ON X.id1 = S.specobjid 
WHERE
    X.ra1 BETWEEN %s and %s and X.dec1 BETWEEN %s and %s
LIMIT 10000
```

### Ejecutar la consulta de instrucción de unión única y los resultados de impresión

```python
df_single = qc.query(sql=query_single,fmt='pandas')
df_single
```

## Ejemplo 2: Una doble unión
Ahora examinaremos tanto los corrimientos al rojo espectroscópicos de SDSS DR17 como la fotometría de LS DR9 escribiendo una consulta con dos declaraciones JOIN. La información relevante que queremos seleccionar de nuestras tres tablas de interés para este ejemplo es:

1. "X" = `sdss_dr17.x1p5__specobj__ls_dr9__tractor`
    - **ra1** (RA del objeto SDSS)
    - **dec1** (Dic del objeto SDSS)
2. "S" = `sdss_dr17.specobj`
    - **z** (desplazamiento al rojo)
3. "L" = `ls_dr9.tractor`
    - **mag_g** (magnitud g convertida)
    - **mag_r** (magnitud r convertida)

### Escriba la consulta de declaración doble JOIN
En este ejemplo utilizamos dos declaraciones JOIN: la primera buscará en la tabla SDSS DR17 `specobj` filas que tengan el mismo valor de identificación de SDSS (`specobjid`) que en la tabla previamente cruzada (`id1`) y recuperará las columnas deseadas de la tabla SDSS DR17 `specobj`. El segundo buscará en la tabla `tractor` de LS DR9 filas que tengan el mismo valor de identificación de LS (`ls_id`) que en la tabla previamente cruzada (`id2`) y recuperará las columnas deseadas de la tabla `tractor` de LS DR9 dentro de la región RA y Dec especificada.

```python
query_double = ("""
SELECT 
    X.ra1 AS ra_sdss, X.dec1 AS dec_sdss,
    S.z,
    L.mag_g, L.mag_r
FROM
    sdss_dr17.x1p5__specobj__ls_dr9__tractor AS X 
JOIN
    sdss_dr17.specobj AS S ON X.id1 = S.specobjid 
JOIN
    ls_dr9.tractor AS L ON X.id2 = L.ls_id
WHERE
    X.ra1 BETWEEN %s and %s and X.dec1 BETWEEN %s and %s
LIMIT 10000
```

### Ejecutar la consulta de la declaración de doble unión y los resultados de impresión

```python
df_double = qc.query(sql=query_double,fmt='pandas')
df_double
```

# Guardar resultados en Vospacio
Vospacio es un espacio de almacenamiento conveniente para que los usuarios guarden su trabajo. Puede almacenar cualquier dato o tipo de archivo. Podemos guardar los resultados de una consulta a nuestro espacio de almacenamiento virtual. Primero, una consulta básica que extrae unos pocos miles de filas de las columnas Specobjid, RA y DEC de la tabla SDSS DR17 Specobj:

```python
basic_query = "SELECT specobjid, ra, dec FROM sdss_dr17.specobj LIMIT 10000"
print(basic_query)
```

##### Enviar la consulta, formatear la salida como CSV y guárdela en Vospacio:

```python
response = qc.query(sql=basic_query,fmt='csv',out='vos://basic_result.csv')
```

##### Asegurémonos de que el archivo se guardara en Vospacio:

```python
sc.ls(name='vos://basic_result.csv')
```

##### Entonces eliminaremos el archivo del vospace:

```python
sc.rm(name='vos://basic_result.csv')
```

##### y asegúrese de que se eliminara:

```python
sc.ls(name='vos://basic_result.csv')
```

# Ejemplo: una tabla de usuario y una tabla previa a la cruzada
Ejemplo: un usuario solo tiene columnas ID, RA y DEC para su propia tabla. Luego realizan un cambio cruzado entre su tabla y 'Gaia_dr3.Gaia_Source`. Al hacerlo, pueden usar la tabla previa a la crujir `gaia_dr3.x1p5__Gaia_source__ls_dr9__tractor` para obtener` ls_dr9.tractor` id, RA y columnas de forma gratuita desde que la tabla pre-crossmated les da la correspondiente `gaia_dr3.Gaia_Source` id. Usaremos la tabla 'Basic_Result` que almacenamos en nuestro MyDB en la sección anterior como nuestra tabla "proporcionada por el usuario".

##### Escribe la consulta CrossMatch de vecino más cercano

```python
query_xmatch = """
SELECT
    b.specobjid AS sdss_id, gg.source_id AS gaia_id, 
    (q3c_dist(b.ra,b.dec,gg.ra,gg.dec)*3600.0) AS dist_arcsec 
FROM
    mydb://basic_result AS b
LEFT JOIN LATERAL (
    SELECT g.* 
    FROM 
        gaia_dr3.gaia_source AS g
    WHERE
        q3c_join(b.ra, b.dec, g.ra, g.dec, 0.01)
    ORDER BY
        q3c_dist(b.ra,b.dec,g.ra,g.dec)
    ASC LIMIT 1
) AS gg ON true
```

##### Envíe la consulta y la salida de CrossMatch a MyDB

```python
df_xmatch = qc.query(sql=query_xmatch,out="mydb://df_xmatch",drop=True) # establecer drop = verdadero para eliminar una tabla ya existente de MyDB con el mismo nombre
```

##### Podemos imprimir la mesa escribiendo una consulta a MyDB

```python
q = "select * from mydb://df_xmatch"
re = qc.query(sql=q,fmt='pandas')
re
```

##### Ahora podemos usar la columna GAIA_ID de nuestra tabla `df_xmatch` arriba para obtener información de objeto` ls_dr9.tractor` utilizando la tabla `gaia_dr3.x1p5__gaia_source__ls_dr9____________

```python
query_1 = """
SELECT
    id2 AS ls_id, ra2 AS ls_ra, dec2 AS ls_dec,
    df_xmatch.gaia_id, df_xmatch.sdss_id
FROM
    gaia_dr3.x1p5__gaia_source__ls_dr9__tractor AS gxl
JOIN
    mydb://df_xmatch AS df_xmatch ON gxl.id1 = df_xmatch.gaia_id
```

##### Envíe la consulta e imprima la tabla resultante

```python
df = qc.query(sql=query_1,fmt='pandas')
df
```
